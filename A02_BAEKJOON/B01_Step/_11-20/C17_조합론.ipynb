```python
# 5	1010	 다리 놓기 (파스칼의 삼각형 DP)
sys.stdin = StringIO('''3
2 2
1 5
13 29''')

import sys
input = sys.stdin.readline

MAX = 30
D = [[0] * (MAX + 1) for _ in range(MAX + 1)]

for n in range(MAX + 1):
    D[n][0] = D[n][n] = 1
    for k in range(1, n):
        D[n][k] = D[n - 1][k - 1] + D[n - 1][k]

for row in D:
    print(row)

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    print(D[M][N])


```
