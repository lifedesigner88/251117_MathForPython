```python
# 2	25192	 인사성 밝은 곰곰이
sys.stdin = StringIO('''9
ENTER
pjshwa
chansol
chogahui05
lms0806
pichulia
r4pidstart
swoon
tony9402''')


import sys
input = sys.stdin.readline

n = int(input())
emoji = set()
count = 0
for _ in range(n):
    cmd = input().rstrip()
    if cmd == 'ENTER':
        emoji.clear()
    else:
        if not cmd in emoji:
            count += 1
            emoji.add(cmd)
print(count)
```
