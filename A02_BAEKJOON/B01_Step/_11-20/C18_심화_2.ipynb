```python
# 2	25192	 인사성 밝은 곰곰이
sys.stdin = StringIO('''9
ENTER
pjshwa
chansol
chogahui05
lms0806
pichulia
r4pidstart
swoon
tony9402''')


import sys
input = sys.stdin.readline

n = int(input())
emoji = set()
count = 0
for _ in range(n):
    cmd = input().rstrip()
    if cmd == 'ENTER':
        emoji.clear()
    else:
        if not cmd in emoji:
            count += 1
            emoji.add(cmd)
print(count)
```

```python
# 3	26069	 붙임성 좋은 총총이
sys.stdin = StringIO('''12
bnb2011 chansol
chansol chogahui05
chogahui05 jthis
jthis ChongChong
jthis jyheo98
jyheo98 lms0806
lms0806 pichulia
pichulia pjshwa
pjshwa r4pidstart
r4pidstart swoon
swoon tony9402
tony9402 bnb2011''')

import sys
input = sys.stdin.readline

n = int(input())

dances = {'ChongChong'}
for _ in range(n):
    a, b = input().split()
    if a in dances or b in dances:
        dances.add(a)
        dances.add(b)
print(len(dances))
```

```python
# 4	2108	 통계학
sys.stdin = StringIO('''3
0
0
-1''')

import sys
from collections import Counter
input = sys.stdin.readline

n = int(input())

nums = [int(input()) for _ in range(n)]
nums.sort()


counts = Counter(nums)
max_value = max(counts.values())
values = [k for k, v in counts.items() if v == max_value]
values.sort()

print(round(sum(nums)/n))
print(nums[n//2])
print(values[0] if len(values) == 1 else values[1])
print(nums[-1] - nums[0])
```

```python
# 5	20920	 영단어 암기는 괴로워
sys.stdin = StringIO('''12 5
appearance
append
attendance
swim
swift
swift
swift
mouse
wallet
mouse
ice
age''')

import sys
from collections import Counter
input = sys.stdin.readline

n, ignore = map(int, input().split())

c = Counter()
for _ in range(n):
    word = input().rstrip()
    if len(word) >= ignore:
        c[word] += 1

lst = sorted(c.items(), key=lambda x: (-x[1], -len(x[0]), x[0]))

for w, _ in lst:
    print(w)

```
