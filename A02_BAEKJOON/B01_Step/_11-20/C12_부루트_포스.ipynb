```python
# 1	2798	 블랙잭
input = StringIO('''10 500
93 181 245 214 315 36 185 138 216 295''').readline

n, target = map(int, input().split())
cards = list(map(int, input().split()))

answer = 0
for i in range(n-2):
    for j in range(i+1, n-1):
        middle_sum = cards[i] + cards[j]
        if middle_sum >= target:
            continue
        for k in range(j+1, n):
            total = middle_sum + cards[k]
            if total <= target and total > answer:
                answer = total
print(answer)


```

```python
# 2	2231	 분해합
input = StringIO('''3''').readline

num = input()
length = (len(num))
num = int(num)
if length > 2:
    start = num-(9*length)
else:
    start = 0
for i in range(start, num):
    digits = [int(digit) for digit in str(i)]
    if i + sum(digits) == num:
        print(i)
        break
else:
    print(0)
```

```python
# 3	19532	 수학은 비대면강의입니다
input = StringIO('''2 5 8 3 -4 -11''').readline

a, b, c, d, e, f = map(int, input().split())
nums = list(range(-999, 999 + 1))
for x in nums:
    for y in nums:
        if a*x + b*y == c and d*x + e*y == f:
            print(x, y)
            break
```

```python
# 4	1018	 체스판 다시 칠하기

input = StringIO('''11 12
BWWBWWBWWBWW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW
BWWBWBWWWBWW
WBWWBWBBWWBW
BWWBWBBWWBWW
WBWWBWBBWWBW''').readline

heigh, width = map(int, input().split())
board = [[w == 'B' for w in input().rstrip()] for _ in range(heigh)]


def check_col8(start_bool, col8):
    counter = 0
    for b in col8:
        if start_bool != b:
            counter += 1
        start_bool = not start_bool
    return counter


def check_row8_col8(row8_col8, minvalue):
    total_counter = [minvalue]
    for start in [True, False]:
        row_counter = 0
        for row8 in row8_col8:
            row_counter += check_col8(start, row8)
            if row_counter >= minvalue:
                break
            start = not start
        total_counter.append(row_counter)
    return min(total_counter)


def extract_col8_row8(start_row, start_col, row_col_board):
    row8_col8 = []
    for row in row_col_board[start_row: start_row + 8]:
        row8_col8.append(row[start_col: start_col + 8])
    return row8_col8


minvalue = 64
for row_idx in range(0, heigh - 8 + 1):
    for col_idx in range(0, width - 8 + 1):
        check_88board = extract_col8_row8(row_idx, col_idx, board)
        minvalue = check_row8_col8(check_88board, minvalue)

print(minvalue)
```
