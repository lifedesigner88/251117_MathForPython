```python
import sys
from io import StringIO
```

```python
# 1	2750	 수 정렬하기
input = StringIO('''5
5
2
3
4
1''').readline

n = int(input())
lst = [int(input().rstrip()) for _ in range(n)]
lst.sort()
for n in lst:
    print(n)
```

```python
# 3	25305	 커트라인
input = StringIO('''5 2
100 76 85 93 98''').readline

n, wins = map(int, input().split())
arr = list(map(int, input().split())) 
arr.sort(reverse=True)
print(arr[wins-1])
```
