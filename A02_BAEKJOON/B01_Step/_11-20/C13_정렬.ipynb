```python
# 5	10989	 수 정렬하기 3
sys.stdin = StringIO('''10
5
2
3
1
4
2
3
5
1
7''')

import sys
input = sys.stdin.readline
n = int(input())

num_counter = [0]*10001

for _ in range(n):
    num_counter[int(input())] += 1

for n in range(1, 10001):
    for _ in range(num_counter[n]):
        print(n)
# 배열로 카운팅 하는 방식으로 풀이 
```

```python
# 7	11650	 좌표 정렬하기
sys.stdin = StringIO('''5
3 4
1 1
1 -1
2 2
3 3''')

import sys
input = sys.stdin.readline

n = int(input())
points = [tuple(map(int, input().split())) for _ in range(n)]
points.sort(key=lambda p: (p[0], p[1]))
for tu in points:
    print(tu[0], tu[1])
```

```python
# 9	1181	 단어 정렬
sys.stdin = StringIO('''13
but
i
wont
hesitate
no
more
no
more
it
cannot
wait
im
yours''')

import sys
input = sys.stdin.readline

n = int(input())

words = {input().rstrip() for _ in range(n)}
words = list(words)
words.sort(key=lambda w: (len(w), w))
for word in words:
    print(word)
```

```python
# 10	10814	 나이순 정렬
sys.stdin = StringIO('''3
21 Junkyu
21 Dohyun
20 Sunyoung''')

import sys
input = sys.stdin.readline

n = int(input())
members = [tuple(input().split()) for _ in range(n)]
members.sort(key=lambda member: (int(member[0])))
for age, name in members:
    print(age, name)
```

```python
# 11	18870	 좌표 압축
sys.stdin = StringIO('''5
2 4 -10 4 -9''')


import sys
from bisect import bisect_left
input = sys.stdin.readline

n = int(input())
nums = list(map(int, input().split()))
sorted_keys = sorted(set(nums))

answer = [str(bisect_left(sorted_keys, num)) for num in nums]
print(" ".join(answer))

```
