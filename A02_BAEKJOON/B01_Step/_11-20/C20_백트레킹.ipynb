```python
# 2	15650	 N과 M (2)
sys.stdin = StringIO('''8 3''')

import sys
input = sys.stdin.readline

n, m = map(int, input().split())

seq = [0] * m

def dfs(depth: int, start: int):
    if depth == m:
        print(*seq)
        return

    # 남은 자리 수
    remain = m - depth
    
    # range 최댓값을 줄여서 불필요한 i를 아예 반복하지 않도록 한다
    # i 는 최대 n - remain + 1 까지 가능해야 남은 자리를 채울 수 있음
    end = n - remain + 2   # range 특성상 +2

    for num in range(start, end):
        seq[depth] = num
        dfs(depth + 1, num + 1)

dfs(0, 1)
```

```python
# 5	9663	 N-Queen

sys.stdin = StringIO('''9''')

import sys
input = sys.stdin.readline

n = int(input())

board = [['.'] * n for _ in range(n)]

q_position = []  
answer = 0
answers = []

def mark_queen(x: int, y: int) -> None:
    # 가로
    for yy in range(n):
        if board[x][yy] == '.':
            board[x][yy] = '*'

    for xx in range(n):
        if board[xx][y] == '.':
            board[xx][y] = '*'

    for a in range(n):
        b1 = (a - x) + y
        b2 = -(a - x) + y
        if 0 <= b1 < n and board[a][b1] == '.':
            board[a][b1] = '*'
        if 0 <= b2 < n and board[a][b2] == '.':
            board[a][b2] = '*'

    board[x][y] = 'Q'

def can_put(x: int, y: int) -> bool:
    return board[x][y] == '.'


def rebuild_board() -> None:
    global board
    board = [['.'] * n for _ in range(n)]
    for x, y in q_position:
        mark_queen(x, y)

from copy import deepcopy
def dfs(row: int) -> None:
    global answer

    if row == n:
        if len(q_position) == n:
            answer += 1
            answers.append(deepcopy(q_position))
        return

    for col in range(n):
        if can_put(row, col):
            q_position.append((row, col))
            mark_queen(row, col)

            dfs(row + 1)

            q_position.pop()
            rebuild_board()


dfs(0)
print(answer)
board = [['.'] * n for _ in range(n)]
for a, b in answers[3]:
    mark_queen(a, b)

for row in board:
    print(*row)
```

```python
# 5	9663	 N-Queen

sys.stdin = StringIO('''14''')

import sys
input = sys.stdin.readline

n = int(input())

col_used=[False] * n

# ↘ 대각선 번호
diag1_used = [False] * (2 * n -1)

# ↙ 대각선 번호
diag2_used = [False] * (2 * n -1)

answer = 0

def dfs(row: int):
    global answer

    if row == n:
        answer +=1
        return 

    for c in range(n):
        d1 = row + c
        d2 = row - c + (n - 1)
        if col_used[c] or diag1_used[d1] or diag2_used[d2]:
            continue
        col_used[c] = diag1_used[d1] = diag2_used[d2] = True
        dfs(row + 1)
        col_used[c] = diag1_used[d1] = diag2_used[d2] = False

dfs(0)
print(answer)
```

```python
# 5	9663	 N-Queen

sys.stdin = StringIO('''14''')

import sys
input = sys.stdin.readline

n = int(input())

answer = 0
full = (1 << n) - 1  # 하위 n비트가 1인 값 (가능한 열 비트 전체)

def dfs(cols: int, diag1: int, diag2: int) -> None:
    """
    cols  : 이미 사용한 열 비트
    diag1 : ↘ 방향 대각선 사용자 비트 (row+col)
    diag2 : ↙ 방향 대각선 사용자 비트 (row-col+n-1 형태를 비트로 반영)
    """
    global answer

    # 열 비트가 n개 모두 채워졌으면, 즉 n개의 퀸을 둔 것
    if cols == full:
        answer += 1
        return

    # 현재 행에서 놓을 수 있는 모든 위치: 아직 공격 안 받은 열들
    # ~(cols | diag1 | diag2) : 사용 불가 칸을 뒤집어서 사용 가능 칸으로
    # & full                  : 상위 불필요 비트 제거
    avail = full & ~(cols | diag1 | diag2)

    # avail에 1이 서 있는 비트(열)마다 퀸을 하나씩 두고 재귀
    while avail:
        # 가장 오른쪽 1비트만 뽑기 (그 위치에 퀸 두기)
        bit = avail & -avail
        avail -= bit  # 그 비트를 제거하고 나머지 위치들 계속 탐색

        dfs(
            cols  | bit,          # 이 열 사용
            (diag1 | bit) << 1,   # ↘ 대각선은 다음 행에서 왼쪽으로 한 칸 밀림
            (diag2 | bit) >> 1    # ↙ 대각선은 다음 행에서 오른쪽으로 한 칸 밀림
        )

dfs(0, 0, 0)
print(answer)
```

```python
# 5	9663	 N-Queen

sys.stdin = StringIO('''14''')

import sys
input = sys.stdin.readline

n = int(input())

answer = 0
full = (1 << n) - 1  # 하위 n비트가 1 (모든 열을 다 쓴 상태)

def dfs(row: int, cols: int, diag1: int, diag2: int) -> None:
    """
    row   : 현재 놓을 행 번호
    cols  : 이미 사용한 열 비트
    diag1 : ↘ 방향 대각선 사용 비트 (다음 행에서 한 칸 왼쪽으로 밀려감)
    diag2 : ↙ 방향 대각선 사용 비트 (다음 행에서 한 칸 오른쪽으로 밀려감)
    """
    global answer

    if row == n:          # n개의 행에 퀸을 다 놓음
        answer += 1
        return

    # 현재 행에서 가능한 열 비트 집합
    avail = full & ~(cols | diag1 | diag2)

    while avail:
        bit = avail & -avail  # 가장 오른쪽 1비트
        avail -= bit          # 그 비트 제거

        dfs(
            row + 1,
            cols | bit,
            (diag1 | bit) << 1,
            (diag2 | bit) >> 1
        )

# 1) 첫 행에서 왼쪽 절반 열만 탐색
half = n // 2
for col in range(half):
    bit = 1 << col
    dfs(
        1,                # 다음은 row 1
        bit,              # 이 열 사용 중
        bit << 1,         # ↘ 대각선
        bit >> 1          # ↙ 대각선
    )

answer *= 2  # 왼쪽 절반 해를 오른쪽 대칭까지 포함

# 2) n이 홀수면, 가운데 열은 따로 한 번만 탐색 (대칭 X)
if n % 2 == 1:
    col = half
    bit = 1 << col
    dfs(
        1,
        bit,
        bit << 1,
        bit >> 1
    )

print(answer)
```

```python
# 7	14888	 연산자 끼워넣기
sys.stdin = StringIO('''6
1 2 3 4 5 6
2 1 1 1''')

import sys
input = sys.stdin.readline

n = int(input())
nums = tuple(map(int, input().split()))
oper = tuple(map(int, input().split()))

def plus(a, b): return a + b
def minus(a, b) : return a - b
def multi(a, b) : return a * b
def div(a, b) : return int(a / b)
funs = [plus, minus, multi, div]

opers = [ funs[i] for i, val in enumerate(oper) for _ in range(val) ]
oper_flag = [True] * (n - 1) 


result = [nums[0]]

answers = []
def dfs(depth:int):
# depth는 연산할 인덱스 (총 n - 1개) (index 는 0 ~ n - 2)

    if depth == n - 1: 
        answers.append(result[depth])
        return
    
    for i in range(n - 1):
        if oper_flag[i]:
            function = opers[i]
            oper_flag[i] = False
            result.append(function(result[depth], nums[depth + 1]))
            dfs(depth + 1)
            oper_flag[i] = True
            result.pop()
dfs(0)
print(max(answers))
print(min(answers))
```

```python
import sys
from io import StringIO

# 테스트용 입력 (백준 제출 시 아래 두 줄은 제거)
sys.stdin = StringIO('''6
1 2 3 4 5 6
2 1 1 1''')

input = sys.stdin.readline

n = int(input())
nums = tuple(map(int, input().split()))
oper = tuple(map(int, input().split()))  # +, -, *, / 개수

def plus(a, b): 
    return a + b

def minus(a, b): 
    return a - b

def multi(a, b): 
    return a * b

def div(a, b):
    # 문제 조건: 0 쪽으로 버림
    if a < 0:
        return -(-a // b)
    return a // b

funs = (plus, minus, multi, div)

# 연산자 목록 펼치기
opers = [funs[i] for i, cnt in enumerate(oper) for _ in range(cnt)]
oper_flag = [True] * len(opers)   # 어떤 연산자를 썼는지 표시

result_stack = [nums[0]]          # 현재까지 계산 결과 스택
answers = []

def dfs(depth: int):
    # depth: 지금까지 사용한 연산자 개수 (0 ~ n-2)

    if depth == n - 1:
        answers.append(result_stack[-1])
        return
    
    for i in range(len(opers)):
        if not oper_flag[i]:
            continue

        oper_flag[i] = False
        cur = result_stack[-1]
        next_val = opers[i](cur, nums[depth + 1])
        result_stack.append(next_val)

        dfs(depth + 1)

        result_stack.pop()
        oper_flag[i] = True

dfs(0)
print(max(answers))
print(min(answers))
```

```python
# 7	14888	 연산자 끼워넣기
sys.stdin = StringIO('''6
1 2 3 4 5 6
2 1 1 1''')

import sys
input = sys.stdin.readline

n = int(input())
nums = tuple(map(int, input().split()))
add, sub, mul, div = map(int, input().split())

MAX = -10**12
MIN = 10**12

def dfs(idx, cur, add, sub, mul, div):
    global MAX, MIN

    if idx == n:
        MAX = max(MAX, cur)
        MIN = min(MIN, cur)
        return 
    
    if add > 0:
        dfs(idx + 1, cur + nums[idx], add -1, sub, mul, div)
    if sub > 0:
        dfs(idx + 1, cur - nums[idx], add, sub - 1, mul, div)
    if mul > 0:
        dfs(idx + 1, cur * nums[idx], add, sub, mul - 1, div)
    if div > 0:
        dfs(idx + 1, int(cur / nums[idx]), add, sub, mul, div - 1)

dfs(1, nums[0], add, sub, mul, div)

print(MAX)
print(MIN)
```

```python
# 8	14889	 스타트와 링크
import sys 

from io import StringIO
sys.stdin = StringIO('''8
0 5 4 5 4 5 4 5
4 0 5 1 2 3 4 5
9 8 0 1 2 3 1 2
9 9 9 0 9 9 9 9
1 1 1 1 0 1 1 1
8 7 6 5 4 0 3 2
9 1 9 1 9 1 0 9
6 5 4 3 2 1 9 0''')

input = sys.stdin.readline

N = int(input())
MIN = 100*20*20
SCORES = [tuple(map(int, input().split())) for _ in range(N)]



def get_team_scores(teams : list[int]):
    # 팀 점수 게산

    team_score = 0
    for i in teams:
        for j in teams:
            if i != j:
                team_score += SCORES[i][j]
    return team_score



# 1명을 뽑아놓고 나머지 4명을 뽑는다. 그리고 남은 5명과 점수 비교. 
# 0 번 선수를 제외하고 나머지 선수들

first_team = list([0])
MEMBERS = tuple(range(1, N)) 

def get_second_team(first_team : list[int]):
    return [m for m in MEMBERS if m not in first_team]

def dfs(need_member:int, start:int):
    # need_member : 뽑아야할 인원

    global MIN
    if need_member == 0 :
        second_team = get_second_team(first_team)
        first_score = get_team_scores(first_team)
        second_score = get_team_scores(second_team)
        diff = abs(first_score - second_score)

        if diff < MIN:
            MIN = diff
        return

    for i in range(start, N - 1 - (need_member - 1)):
        first_team.append(MEMBERS[i])
        dfs(need_member - 1, i + 1)
        first_team.pop()


dfs(N//2 - 1, 0)

print(MIN)
```

```python
# 8	14889	 스타트와 링크
import sys

from io import StringIO
sys.stdin = StringIO('''6
0 1 2 3 4 5
1 0 2 3 4 5
1 2 0 3 4 5
1 2 3 0 4 5
1 2 3 4 0 5
1 2 3 4 5 0''')

input = sys.stdin.readline

N = int(input())
S = [tuple(map(int, input().split())) for _ in range(N)]

INF = 10**9
answer = INF
visited = [False] * N


def cal_diff():
    start_team = list()
    link_team = list()
    for i in range(N):
        if visited[i]:
            start_team.append(i)
        else:
            link_team.append(i)

    start_score = 0
    link_score = 0

    for i in range(N // 2):
        for j in range(i + 1, N // 2):
            a, b = start_team[i], start_team[j]
            c, d = link_team[i], link_team[j]
            start_score += S[a][b] + S[b][a]
            link_score += S[c][d] + S[d][c]

    return abs(start_score - link_score)


def dfs(idx, cnt):
    global answer

    if cnt == N // 2:
        answer = min(answer, cal_diff())
        return

    for i in range(idx, N):
        if not visited[i]:
            visited[i] = True
            dfs(i + 1, cnt + 1)
            visited[i] = False

visited[0] = True

dfs(1, 1)
print(answer)
```
