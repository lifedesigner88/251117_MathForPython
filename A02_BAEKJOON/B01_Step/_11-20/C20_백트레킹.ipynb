```python
# 1	15649	 N과 M (1)
sys.stdin = StringIO('''4 4''')

import sys
input = sys.stdin.readline

n, m = map(int, input().split())

path = []
used = [False] * (n + 1)

def dfs(depth):
    if depth == m:
        print(*path)
        return
    for num in range(1, n+1):
        if not used[num]:
            used[num] = True
            path.append(num)

            dfs(depth + 1)

            path.pop()
            used[num] = False
dfs(0)
            
```
