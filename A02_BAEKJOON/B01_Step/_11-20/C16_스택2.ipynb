```python
# 1	28278	 스택 2
sys.stdin = StringIO('''9
4
1 3
1 5
3
2
5
2
2
5''')

import sys
input = sys.stdin.readline

n = int(input())

stack = []

for _ in range(n):
    c = tuple(map(int, input().split()))

    if c[0] == 1:
        stack.append(c[1])
    elif c[0] == 2:
        if stack:
            print(stack.pop())
        else:
            print(-1)
    elif c[0] == 3:
        print(len(stack))
    elif c[0] == 4:
        if stack:
            print(0)
        else:
            print(1)
    else:
        if stack:
            print(stack[-1])
        else:
            print(-1)
```

```python
# 3	9012	 괄호
sys.stdin = StringIO('''6
(())())
(((()())()
(()())((()))
((()()(()))(((())))()
()()()()(()()())()
(()((())()(''')

import sys
input = sys.stdin.readline

n = int(input())

for _ in range(n):
    stack = list()
    for p in input().rstrip():
        if p == '(':
            stack.append(p)
        else:
            if stack and stack[-1] == '(': 
                stack.pop()
            else:
                stack.append(p)
                break
    if stack:
        print('NO')
    else: 
        print('YES')

```

```python
# 4	4949	 균형잡힌 세상
sys.stdin = StringIO('''So when I die (the [first] I will see in (heaven) is a score list).
[ first in ] ( first out ).
Half Moon tonight (At least it is better than no Moon at all].
A rope may form )( a trail in a maze.
Help( I[m being held prisoner in a fortune cookie factory)].
([ (([( [ ] ) ( ) (( ))] )) ]).
 .
.''')

import sys
input = sys.stdin.readline

while True:
    sentance = input().rstrip()
    if sentance == '.' : break

    stack = list()
    for c in sentance:
        if c == '(':
            stack.append(c)   
        elif c == ')':
            if stack and stack[-1] == '(':
                stack.pop()
            else:
                stack.append(c)
                break
        
        elif c == '[':
            stack.append(c)
        elif c == ']':
            if stack and stack[-1] == '[':
                stack.pop()
            else:
                stack.append(c)
                break
    print('no' if stack else 'yes')
```

```python
# 5	12789	 도키도키 간식드리미
sys.stdin = StringIO('''5
5 4 1 3 2''')

import sys
input = sys.stdin.readline

n = int(input())
line = list(map(int, input().split()))
line.reverse()

stack = list()

for order in range(1, n+1):
    if line and order == line[-1]:
        line.pop()
        continue
    elif stack and order == stack[-1]:
        stack.pop()
        continue
    else:
        while line and order != line[-1]:
            stack.append(line.pop())
        if not line:
            print('Sad')
            break
        else:
            line.pop()
            continue
else:
    print('Nice')


```

```python
# 6	18258	 큐 2
sys.stdin = StringIO('''15
push 1
push 2
front
back
size
empty
pop
pop
pop
size
empty
pop
push 3
empty
front''')

import sys
from collections import deque
input = sys.stdin.readline

n = int(input())

dq = deque()
for _ in range(n):
    c = input().split()
    if c[0] == 'push':
        dq.append(c[1])
    elif c[0] == 'pop':
        print(dq.popleft() if dq else -1)
    elif c[0] == 'size':
        print(len(dq))
    elif c[0] == 'empty':
        print(0 if dq else 1)
    elif c[0] == 'front':
        print(dq[0] if dq else -1)
    else :
        print(dq[-1] if dq else -1)


```
