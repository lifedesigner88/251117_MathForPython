```python
# 1	28278	 스택 2
sys.stdin = StringIO('''9
4
1 3
1 5
3
2
5
2
2
5''')

import sys
input = sys.stdin.readline

n = int(input())

stack = []

for _ in range(n):
    c = tuple(map(int, input().split()))

    if c[0] == 1:
        stack.append(c[1])
    elif c[0] == 2:
        if stack:
            print(stack.pop())
        else:
            print(-1)
    elif c[0] == 3:
        print(len(stack))
    elif c[0] == 4:
        if stack:
            print(0)
        else:
            print(1)
    else:
        if stack:
            print(stack[-1])
        else:
            print(-1)
```

```python
# 3	9012	 괄호
sys.stdin = StringIO('''6
(())())
(((()())()
(()())((()))
((()()(()))(((())))()
()()()()(()()())()
(()((())()(''')

import sys
input = sys.stdin.readline

n = int(input())

for _ in range(n):
    stack = list()
    for p in input().rstrip():
        if p == '(':
            stack.append(p)
        else:
            if stack and stack[-1] == '(': 
                stack.pop()
            else:
                stack.append(p)
                break
    if stack:
        print('NO')
    else: 
        print('YES')

```

```python
# 4	4949	 균형잡힌 세상
sys.stdin = StringIO('''So when I die (the [first] I will see in (heaven) is a score list).
[ first in ] ( first out ).
Half Moon tonight (At least it is better than no Moon at all].
A rope may form )( a trail in a maze.
Help( I[m being held prisoner in a fortune cookie factory)].
([ (([( [ ] ) ( ) (( ))] )) ]).
 .
.''')

import sys
input = sys.stdin.readline

while True:
    sentance = input().rstrip()
    if sentance == '.' : break

    stack = list()
    for c in sentance:
        if c == '(':
            stack.append(c)   
        elif c == ')':
            if stack and stack[-1] == '(':
                stack.pop()
            else:
                stack.append(c)
                break
        
        elif c == '[':
            stack.append(c)
        elif c == ']':
            if stack and stack[-1] == '[':
                stack.pop()
            else:
                stack.append(c)
                break
    print('no' if stack else 'yes')
```

```python
# 5	12789	 도키도키 간식드리미
sys.stdin = StringIO('''5
5 4 1 3 2''')

import sys
input = sys.stdin.readline

n = int(input())
line = list(map(int, input().split()))
line.reverse()

stack = list()

for order in range(1, n+1):
    if line and order == line[-1]:
        line.pop()
        continue
    elif stack and order == stack[-1]:
        stack.pop()
        continue
    else:
        while line and order != line[-1]:
            stack.append(line.pop())
        if not line:
            print('Sad')
            break
        else:
            line.pop()
            continue
else:
    print('Nice')


```

```python
# 6	18258	 큐 2
sys.stdin = StringIO('''15
push 1
push 2
front
back
size
empty
pop
pop
pop
size
empty
pop
push 3
empty
front''')

import sys
from collections import deque
input = sys.stdin.readline

n = int(input())

dq = deque()
for _ in range(n):
    c = input().split()
    if c[0] == 'push':
        dq.append(c[1])
    elif c[0] == 'pop':
        print(dq.popleft() if dq else -1)
    elif c[0] == 'size':
        print(len(dq))
    elif c[0] == 'empty':
        print(0 if dq else 1)
    elif c[0] == 'front':
        print(dq[0] if dq else -1)
    else :
        print(dq[-1] if dq else -1)


```

```python
# 7	2164	 카드2
sys.stdin = StringIO('''6''')

import sys
from collections import deque
input = sys.stdin.readline

n = int(input())

dq = deque(range(1, n+1))

while len(dq) > 1:
    dq.popleft()
    dq.append(dq.popleft())

print(dq.pop())
```

```python
# 8	11866	 요세푸스 문제 0
sys.stdin = StringIO('''7 3''')

import sys
from collections import deque
input = sys.stdin.readline

n, k = map(int, input().split())
dq = deque(range(1, n + 1))
answer = list()
for _ in range(n):
    dq.rotate(-(k-1))
    answer.append(str(dq.popleft()))
print(f"<{", ".join(answer)}>")
```

```python
# 9	28279	 덱 2

sys.stdin = StringIO('''11
6
1 3
1 8
7
8
3
2 5
1 2
5
4
4''')

import sys
from collections import deque
input = sys.stdin.readline

n = int(input())

dq = deque()

switch = {
    1: lambda x: dq.appendleft(x),
    2: lambda x: dq.append(x),
    3: lambda: dq.popleft() if dq else -1,
    4: lambda: dq.pop() if dq else -1,
    5: lambda: len(dq),
    6: lambda: 0 if dq else 1,
    7: lambda: dq[0] if dq else -1,
    8: lambda: dq[-1] if dq else -1
}

for _ in range(n):
    c = input().split()
    func = switch.get(int(c[0]))

    if len(c)==2:
        func(int(c[1]))
    else:
        print(func())
        


```

```python
# 10	2346	 풍선 터뜨리기
sys.stdin = StringIO('''5
3 2 1 -3 -1''')

import sys
from collections import deque
input = sys.stdin.readline

n = int(input())
lst = input().split()
dq = deque((str(i + 1), int(lst[i])) for i in range(n) )

answer = list()
while dq:
    idx, move = dq.popleft()
    answer.append(idx)
    if move > 0:
        dq.rotate(-(move-1))
    else:
        dq.rotate(-move)
print(" ".join(answer))



#  list 기반
sys.stdin = StringIO('''5
3 2 1 -3 -1''')

import sys
input = sys.stdin.readline
n = int(input())
lst = input().split()

answer = []
i = 0
while True:
    move = int(lst[i])
    answer.append(str(i+1))
    if len(answer) == n: break
    lst[i] = "#"
    if move > 0:
        while move > 0:
            i = (i+1)%n      
            if lst[i] != "#":
                move -= 1
    else :
        while move < 0:
            i = (i+(n-1))%n
            if lst[i] != "#":
                move += 1
print(" ".join(answer))
```

```python
# 11	24511	 queuestack
sys.stdin = StringIO('''4
0 1 1 0
1 2 3 4
3
2 4 7''')

import sys
input = sys.stdin.readline

n = int(input())

que_stack = input().split()
start_ins = input().split()

m = int(input())
appends = input().split()

result = [start_ins[i] for i in range(n) if que_stack[i] == '0']

result.reverse()
result.extend(appends)
print(" ".join(result[:m]))
```
