```python
# 1	27433	 팩토리얼 2
sys.stdin = StringIO('''10''')

import sys
input = sys.stdin.readline

n = int(input())

def factorial(num):
    if num < 1: return 1
    return num * factorial(num-1)

print(factorial(n))
```

```python
# 2	10870	 피보나치 수 5

sys.stdin = StringIO('''10''')

import sys
input = sys.stdin.readline

n = int(input())

def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibo(n - 1) + fibo(n-2)

print(fibo(n))
```

```python
# 3	25501	 재귀의 귀재
sys.stdin = StringIO('''5
AAA
ABBA
ABABA
ABCA
PALINDROME''')


import sys
input = sys.stdin.readline

def recursion(s, l, r):
    recursion.count += 1
    if l >= r: return 1
    elif s[l] != s[r]: return 0
    else: return recursion(s, l+1, r-1)

def isPalindrome(s):
    recursion.count = 0
    return recursion(s, 0, len(s)-1), recursion.count

n = int(input())

for _ in range(n):
    p, c = isPalindrome(input().rstrip())
    print(p, c)
```

```python
# 4	24060	 알고리즘 수업 - 병합 정렬 1
sys.stdin = StringIO('''5 7
4 5 1 3 2''')

import sys
input = sys.stdin.readline

n, k = map(int, input().split())
nums = list(map(int, input().split()))

cnt = 0

class Found(Exception):
    def __init__(self, value):
        self.value = value

def merge_sort(input_lst, start, end):
    if start < end:
        middle = (end + start)//2
        merge_sort(input_lst, start, middle)
        merge_sort(input_lst, middle + 1, end)
        merge(input_lst, start, middle, end)

def merge(origin, start, middle, end):
    global cnt

    i, j = start, middle + 1
    temp = list()
    while i <= middle and j <= end:
        if origin[i] <= origin[j]:
            temp.append(origin[i])
            i += 1
        else:
            temp.append(origin[j])
            j += 1
    if i <= middle:
        temp.extend(origin[i: middle+1])
    if j <= end:
        temp.extend(origin[j: end+1])

    t_idx = 0
    for idx in range(start, end + 1):
        origin[idx] = temp[t_idx]
        cnt += 1
        if cnt == k:
            raise Found(temp[t_idx])
        t_idx += 1

try:
    merge_sort(nums, 0, len(nums)-1)
    print(-1)
except Found as f:
    print(f.value)
```

```python
# 5	4779	 칸토어 집합
sys.stdin = StringIO('''0
1
3
4''')


import sys
input = sys.stdin.readline

def mark_blank(array, start, end):
    if start == end: return
    term = (end - start + 1) // 3
    blank_start = start + term
    blank_end = start + 2*term
    array[blank_start:blank_end] = [False]*term

    mark_blank(array, start, blank_start-1)
    mark_blank(array, blank_end, end)

while True:
    n = input().rstrip()
    if not n : break
    n = int(n)
    result = [True]* 3**n
    mark_blank(result, 0, len(result) -1)
    answer = ['-' if i else ' ' for i in result ]
    print("".join(answer))
```

```python
# 6	2447	 별 찍기 - 10

sys.stdin = StringIO('''27''')

import sys
input = sys.stdin.readline

def return3point(start, end):
    term = (end - start + 1) // 3
    middle_start = start + term
    middle_end = start + 2*term
    return (start, middle_start), (middle_start, middle_end), (middle_end, end+1) 

def fill_star(n):
    if n == 1:
        return ['*']

    sub_leg = n//3 
    sub_array = fill_star(sub_leg)

    array = [[" "] * n for _ in range(n)]
    points = return3point(0, n - 1)
    for i , (col_start, col_end) in enumerate(points):
        for j, (row_start, row_end) in enumerate(points):
            if i==j==1: 
                continue
            for col in range(col_start, col_end):
                for row in range(row_start, row_end):
                    array[col][row] = sub_array[col%sub_leg][row%sub_leg]
    return array

n = int(input())
for row in fill_star(n):
    print("".join(row))
```

```python
# 6	2447	 별 찍기 - 10

sys.stdin = StringIO('''9''')

import sys
input = sys.stdin.readline

def fill_star(n):
    # 항상 2D 배열을 반환하도록 통일
    if n == 1:
        return [['*']]

    sub_n = n // 3
    sub = fill_star(sub_n)                  # sub_n x sub_n 패턴
    arr = [[' '] * n for _ in range(n)]     # 전체 n x n 배열

    # 3x3 블록 순회
    for bi in range(3):         # block row index (0,1,2)
        for bj in range(3):     # block col index (0,1,2)
            # 가운데 블록은 비워둠
            if bi == 1 and bj == 1:
                continue

            start_r = bi * sub_n
            start_c = bj * sub_n

            # sub 패턴을 해당 블록 위치에 복사
            for i in range(sub_n):
                for j in range(sub_n):
                    arr[start_r + i][start_c + j] = sub[i][j]

    return arr

n = int(input())
board = fill_star(n)
for row in board:
    print("".join(row))
```

```python
# 6	2447	 별 찍기 - 10

sys.stdin = StringIO('''27''')

import sys
input = sys.stdin.readline

def solve(n):
    if n == 1:
        return ["*"]
    
    sub = solve(n//3)
    lines = list()

    for s in sub:
        lines.append(s * 3)
    
    space = " " * (n//3)
    for s in sub:
        lines.append(s + space + s)
    
    for s in sub:
        lines.append(s * 3)

    return lines

n = int(input())
print("\n".join(solve(n)))
```

```python
# 7	11729	 하노이 탑 이동 순서

sys.stdin = StringIO('''3''')

import sys
input = sys.stdin.readline
num = int(input())

def solution(n, start, middle, end):
    if n == 0:
        return []

    answer = []
    befor_sub_sol = solution(n-1, start, end, middle)
    answer.extend(befor_sub_sol)

    answer.append((start, end))

    after_sub_sol = solution(n-1, middle, start, end)
    answer.extend(after_sub_sol)

    return answer

moves = solution(num, 1, 2, 3)
print(len(moves))

for a, b in moves:
    print(a, b)

```

```python
# 7	11729	 하노이 탑 이동 순서

sys.stdin = StringIO('''3''')
import sys
input = sys.stdin.readline
write = sys.stdout.write

n = int(input())
write(str(2**n - 1) + "\n")

def hanoi(k, start, middle, end):
    if k == 0:
        return
    hanoi(k - 1, start, end, middle)
    write(f"{start} {end}\n")
    hanoi(k - 1, middle, start, end)

hanoi(n, 1, 2, 3)
```
