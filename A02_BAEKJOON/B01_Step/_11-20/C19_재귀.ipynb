```python
# 1	27433	 팩토리얼 2
sys.stdin = StringIO('''10''')

import sys
input = sys.stdin.readline

n = int(input())

def factorial(num):
    if num < 1: return 1
    return num * factorial(num-1)

print(factorial(n))
```

```python
# 2	10870	 피보나치 수 5

sys.stdin = StringIO('''10''')

import sys
input = sys.stdin.readline

n = int(input())

def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibo(n - 1) + fibo(n-2)

print(fibo(n))
```

```python
# 3	25501	 재귀의 귀재
sys.stdin = StringIO('''5
AAA
ABBA
ABABA
ABCA
PALINDROME''')


import sys
input = sys.stdin.readline

def recursion(s, l, r):
    recursion.count += 1
    if l >= r: return 1
    elif s[l] != s[r]: return 0
    else: return recursion(s, l+1, r-1)

def isPalindrome(s):
    recursion.count = 0
    return recursion(s, 0, len(s)-1), recursion.count

n = int(input())

for _ in range(n):
    p, c = isPalindrome(input().rstrip())
    print(p, c)
```
