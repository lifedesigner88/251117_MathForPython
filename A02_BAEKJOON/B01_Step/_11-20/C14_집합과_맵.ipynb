```python
# 1	10815	 숫자 카드
sys.stdin = StringIO('''5
6 3 2 10 -10
8
10 9 -5 2 3 4 5 -10''')

import sys
input = sys.stdin.readline

input()
cards = set(map(int, input().split()))
input()
nums = map(int, input().split())
answer = ["1" if num in cards else "0" for num in nums]
print(" ".join(answer))
```

```python
# 2	14425	 문자열 집합
sys.stdin = StringIO('''5 11
baekjoononlinejudge
startlink
codeplus
sundaycoding
codingsh
baekjoon
codeplus
codeminus
startlink
starlink
sundaycoding
codingsh
codinghs
sondaycoding
startrink
icerink''')

import sys
input = sys.stdin.readline
n, m = map(int, input().split())

words_set = set(input().rstrip() for _ in range(n))
count = 0
for _ in range(m):
    if input().rstrip() in words_set:
        count += 1
print(count)
```

```python
# 3	7785	 회사에 있는 사람
sys.stdin = StringIO('''4
Baha enter
Askar enter
Baha leave
Artem enter''')

import sys
input = sys.stdin.readline

n = int(input())
members = set()
for _ in range(n):
    name, info = input().split()
    if info[0] == 'e':
        members.add(name)
    else:
        members.remove(name)

for n in sorted(members, reverse=True):
    print(n)
```

```python
# 4	1620	 나는야 포켓몬 마스터 이다솜
sys.stdin = StringIO('''26 5
Bulbasaur
Ivysaur
Venusaur
Charmander
Charmeleon
Charizard
Squirtle
Wartortle
Blastoise
Caterpie
Metapod
Butterfree
Weedle
Kakuna
Beedrill
Pidgey
Pidgeotto
Pidgeot
Rattata
Raticate
Spearow
Fearow
Ekans
Arbok
Pikachu
Raichu
25
Raichu
3
Pidgey
Kakuna''')

import sys
input = sys.stdin.readline

m, p = map(int, input().split())

key_to_num = {}
num_to_key = {}

for i in range(m):
    key = input().rstrip()
    num = i + 1
    key_to_num[key] = num
    num_to_key[num] = key

for _ in range(p):
    temp = input().rstrip()
    if temp.isdigit():
        print(num_to_key[int(temp)])
    else:
        print(key_to_num[temp])
```

```python
# 5	10816	 숫자 카드 2
sys.stdin = StringIO('''10
6 3 2 10 10 10 -10 -10 7 3
8
10 9 -5 2 3 4 5 -10''')

import sys
rl = sys.stdin.readline
rl()

cards = rl().split()
card_dict = {}
for card in cards:
    card_dict[card] = card_dict.get(card, 0) + 1

rl()
checks = rl().split()
answer = (str(card_dict.get(check, 0)) for check in checks)
print(" ".join(answer))
```

```python
# 7	1269	 대칭 차집합
sys.stdin = StringIO('''3 5
1 2 4
2 3 4 5 6''')

import sys
rl = sys.stdin.readline

a, b = map(int, rl().split())

a_set = set(rl().split())
b_set = set(rl().split())
tran_set = a_set & b_set

print(a + b - 2*len(tran_set))
```

```python
# 8	11478	 서로 다른 부분 문자열의 개수
sys.stdin = StringIO('''ababc''')

import sys
rl = sys.stdin.readline

word = rl().rstrip()
n = len(word)
answer = set()
for length in range(1, n+1):        
    for start in range(n - length + 1):  
        answer.add(word[start:start+length])
print(len(answer))
```
