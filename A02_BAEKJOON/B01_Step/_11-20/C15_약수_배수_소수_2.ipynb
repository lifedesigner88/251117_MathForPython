```python
# 3	1735	 분수 합
sys.stdin = StringIO('''2 7
3 5''')


import sys
from math import gcd
input = sys.stdin.readline

fracs = [tuple(map(int, input().split())) for _ in range(2)]
numer = fracs[0][0]*fracs[1][1] + fracs[1][0]*fracs[0][1]
denom = fracs[0][1]*fracs[1][1]
gcd_v = gcd(numer, denom)
print(numer//gcd_v, denom//gcd_v)
```

```python
# 4	2485	 가로수
sys.stdin = StringIO('''4
1
3
7
13''')


import sys
from math import gcd
input = sys.stdin.readline

n = int(input())
start = int(input())
prev = start
terms = set()
for _ in range(n-1):
    position = int(input())
    terms.add(position - prev)
    prev = position
last_p = position

length_terms = len(terms)
terms_tuple = tuple(terms)
final_gcd = terms_tuple[0]
if len(terms) > 1:
    for i in range(1, len(terms_tuple)):
        final_gcd = gcd(final_gcd, terms_tuple[i])

print((last_p - start)//final_gcd + 1 - n)
```

```python
# 5	4134	 다음 소수
sys.stdin = StringIO('''3
6
20
100''')

import sys
from math import isqrt
input = sys.stdin.readline

def isprime(num):
    if num == 0 or num == 1 :
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for div in range(3, isqrt(num) + 1, 2):
        if num % div == 0:
            return False
    return True 

n = int(input())
for _ in range(n):
    number = int(input())
    while True:
        if isprime(number):
            print(number)
            break
        number += 1


```

```python
# 6	1929	 소수 구하기

sys.stdin = StringIO('''3 16''')

import sys
from math import isqrt
input = sys.stdin.readline

start, end = map(int, input().split())

flag = [True]*(end + 1)
flag[0]=flag[1]=False

for i in range(2, isqrt(end) + 1):
    if flag[i]:
        for num in range(i*i, end + 1, i):
            flag[num] = False
for check in range(start, end + 1):
    if flag[check]:
        print(check)
```

```python
# 7	4948	 베르트랑 공준
sys.stdin = StringIO('''1
10
13
100
1000
10000
100000
0''')

import sys
from math import isqrt

input = sys.stdin.readline

max_value = 123456*2
is_prime = [True] * (max_value + 1)
is_prime[0] = is_prime[1] = False

for i in range(2, isqrt(max_value) + 1):
    if is_prime[i]:
        for div in range(i*i, max_value + 1, i):
            is_prime[div] = False

prime_count = [0] * (max_value + 1)
for i in range(max_value + 1):
    prime_count[i] = prime_count[i-1] + (1 if is_prime[i] else 0)

while True :
    num = int(input())
    if num == 0:
        break
    print(prime_count[2*num] - prime_count[num])
```

```python
# 8	17103	 골드바흐 파티션

sys.stdin = StringIO('''5
6
8
10
12
100''')

import sys
from math import isqrt
input = sys.stdin.readline

max_value = 10**6
is_prime = [True]*(max_value + 1)
is_prime[0] = is_prime[1] = False

for i in range(2, isqrt(max_value) + 1):
    if is_prime[i]:
        for div in range(i*i, max_value + 1, i):
            is_prime[div] = False

primes = [i for i, val in enumerate(is_prime) if val]

n = int(input())
for _ in range(n):
    num = int(input())
    count = 0
    for p in primes:
        if p > num // 2:
            break
        if is_prime[num - p]:
            count += 1
    print(count)
```
