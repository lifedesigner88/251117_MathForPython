```python
# 2. 10950 A+B - 3
t = int(input())

for _ in range(t):
    a, b = map(int, input().split())
    print(a + b)
```

```python
# 3. 8393 합
t = int(input())

sum = 0
for i in range(1, t + 1):
    sum += i
print(sum)
```

```python
# 4. 25340 영수증

from io import StringIO
input = StringIO("""260000
4
20000 5
30000 2
10000 6
5000 8""").readline

total = int(input())
count = int(input())

for _ in range(count):
    price, num = map(int, input().split())
    total -= price*num

print("Yes" if total==0 else "No")
```

```python
# 5. 25314 코딩은 체육과목 입니다. 
a = int(input())
answer = "int"
for _ in range(a//4):
    answer = "long " + answer
print(answer)
```

```python
# 6. 155552 빠른 A+B
from io import StringIO
inputs = StringIO("""5
1 1
12 34
5 500
40 60
1000 1000
""")

from sys import stdin
stdin = inputs

rl = stdin.readline
n = int(rl().rstrip())
for _ in range(n):
    print(sum(map(int, rl().split())))
```

```python
# 7. 11021 A+B -7

from sys import stdin
from io import StringIO
stdin = StringIO('''5
1 1
2 3
3 4
9 8
5 2''')

rl = stdin.readline
n = int(rl())
for i in range(n):
    print(f"Case #{i+1}:", sum(map(int, rl().split())))
```

```python
# 8. 11022 A+B - 8

from sys import stdin
from io import StringIO
stdin = StringIO("""5
1 1
2 3
3 4
9 8
5 2""")

rl = stdin.readline
n = int(rl())
for i in range(n):
    a, b = map(int, rl().split())
    print(f"Case #{i+1}: {a} + {b} = {a+b}")
```

```python
# 9. 2438 별 찍기 - 1

n = int(input())
for i in range(n):
    print("*" * (i+1))
```

```python
# 10. 2439 별 찍기 - 2

n = int(input())
for i in range(n):
    print(" " * (n - i - 1) + "*" * (i + 1))
```

```python
# 11. A+B - 5

from sys import stdin
from io import StringIO

stdin = StringIO("""1 1
2 3
3 4
9 8
5 2
0 0""")

rl = stdin.readline

while True:
    a , b = map(int, rl().split())
    if a == b == 0 :
        break
    print(a+b)

```

```python
# 12. 10951 A+B - 4

from sys import stdin
from io import StringIO
stdin = StringIO('''1 1
2 3
3 4
9 8
5 2''')

rl = stdin.readline

while True:
    line = rl()
    if not line:
        break
    print(sum(map(int, line.split())))
```
