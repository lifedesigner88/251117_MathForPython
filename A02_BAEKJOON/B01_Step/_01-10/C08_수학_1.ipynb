```python
# 1	2745	진법 변환

sys.stdin = StringIO('''ZZZZZ 36''')

import sys
a, b = sys.stdin.readline().split()
base = int(b)

sum = 0
for i, digit in enumerate(reversed(a)):
    if digit.isdigit():
        value = int(digit)
    else:
        value = ord(digit) - ord('A') + 10
    sum += value * (base**i)
print(sum)
```

```python
# 2	11005	진법 변환 2

sys.stdin = StringIO('''60466175 36''')

import sys
a, b = map(int, sys.stdin.readline().split())

answer = []
while True:
    digit = a % b # a, b = divmod(a, b)
    if digit >= 10:
        digit = chr(digit + ord('Z') - 35)
    else:
        digit = str(digit)
    answer.append(digit)
    a = a//b
    if a == 0:
        break
print("".join(reversed(answer)))
```

```python
# 3	2720	세탁소 사장 동혁
sys.stdin = StringIO('''3
124
25
194''')

import sys
rl = sys.stdin.readline

n = int(rl())
for _ in range(n):
    payback = int(rl())
    answer = []
    for coin in [25, 10, 5, 1]:
        count, payback = divmod(payback, coin)
        answer.append(str(count))
    print(" ".join(answer))
```

```python
# 5	2292	벌집

sys.stdin = StringIO('''13''')

import sys
n = int(sys.stdin.readline())

layer = 1
range_max = 1
while n > range_max:
    range_max += 6 * layer
    layer += 1

print(layer)
```

```python
# 6	1193	분수찾기
sys.stdin = StringIO('''14''')

import sys
n = int(sys.stdin.readline())

round = 1
while True:
    temp = n - round
    if temp <= 0:
        break
    n = temp
    round += 1

if round % 2 == 0:
    print(f"{n}/{round + 1-n} ")
else:
    print(f"{round + 1 - n}/{n}")


'''
1 / n 은 몇번째 인가요? 

1   1       
2   2       +1
3   4       +2
4   7       +3
5   11      +4
6   16      +5



'''
```
