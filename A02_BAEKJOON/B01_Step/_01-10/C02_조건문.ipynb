```python
# 4. 14681 사분면 고르기
x = int(input())
y = int(input())

if x > 0:
    if y > 0:
        print(1)
    else:
        print(4)
else:
    if y > 0:
        print(2)
    else:
        print(3)
```

```python
# 5. 2884 IQ 검사
a, b = map(int, input().split())

if b >= 45:
    print(a, b-45)

else:
    if a != 0:
        print(a-1, b + (60-45))
    else:
        print(23, b + (60-45))
```

```python
# 6. 2525 오븐시계
h, m = map(int, input().split())
t = int(input())

th, tm = [t//60, t%60]

totalh = (h + th) % 24
totalm = m + tm

if totalm < 60:
    print(totalh, totalm)
else:
    if totalh == 23:
        print(0, totalm - 60)
    else:
        print(totalh + 1, totalm - 60)
```

```python
# 7. 2480 주사위 세개
a, b, c = map(int, input().split())

if a==b and b==c:
    print(10000 + a*1000)
elif a==b or a==c:
    print(1000 + a*100)
elif b==c:
    print(1000 + b*100)
else:
    largest = a
    if b > a:
        largest = b
    if c > largest:
        largest = c
    print(largest*100)
```
