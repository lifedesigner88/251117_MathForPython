```python
# 5	10810	공 넣기

stdin = StringIO('''5 4
1 2 3
3 4 4
1 4 1
2 2 2''')

rl = stdin.readline

length, trying = map(int, rl().split())

bucket = [0] * length

for _ in range(trying):
    start, end, number = map(int, rl().split())
    for i in range(start-1, end):
        bucket[i] = number

print(" ".join(map(str, bucket)))
```
