```python
# 5	10810	공 넣기

stdin = StringIO('''5 4
1 2 3
3 4 4
1 4 1
2 2 2''')

rl = stdin.readline

length, trying = map(int, rl().split())

bucket = [0] * length

for _ in range(trying):
    start, end, number = map(int, rl().split())
    for i in range(start-1, end):
        bucket[i] = number

print(" ".join(map(str, bucket)))
```

```python
# 6	10813	공 바꾸기

stdin = StringIO('''5 4
1 2
3 4
1 4
2 2''')

rl = stdin.readline
length, trying = map(int, rl().split())

bucket = list(range(1, length + 1))
for _ in range(trying):
    a, b = map(int, rl().split())
    bucket[a-1], bucket[b-1] = bucket[b-1], bucket[a-1]

print(" ".join(map(str, bucket)))
```

```python
# 9	10811	바구니 뒤집기

stdin = StringIO('''5 4
1 2
3 4
1 4
2 2''')

rl = stdin.readline

length, trying = map(int, rl().split())
buckets = list(range(1, length + 1))

for _ in range(trying):
    a, b = map(int, rl().split())
    buckets[a-1:b] = buckets[a-1:b][::-1]

print(" ".join(map(str, buckets)))
```

```python
# 10	1546	평균

stdin = StringIO('''3
40 80 60''')

rl = stdin.readline

n = int(rl())

score = list(map(int, rl().split()))
max_val = max(score)
new_score = [ (val/max_val)*100 for val in score]
print(sum(new_score)/n)
```
