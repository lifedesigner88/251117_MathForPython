```python
# 1	2738	행렬 덧셈

sys.stdin = StringIO('''3 3
1 1 1
2 2 2
0 1 0
3 3 3
4 4 4
5 5 100''')

import sys
rl = sys.stdin.readline

n, m = map(int, rl().split())
matrix = [list(map(int, rl().split())) for _ in range(n)]

for i in range(n):
    will_added = list(map(int, rl().split()))
    for j in range(m):
        matrix[i][j] += will_added[j]
    print(" ".join(map(str, matrix[i])))

```

```python
# 2	2566	최댓값
sys.stdin = StringIO('''3 23 85 34 17 74 25 52 65
10 7 39 42 88 52 14 72 63
87 42 18 78 53 45 18 84 53
34 28 64 85 12 16 75 36 55
21 77 45 35 28 75 90 76 1
25 87 65 15 28 11 37 28 74
65 27 75 41 7 89 78 64 39
47 47 70 45 23 65 3 41 44
87 13 82 38 31 12 29 29 80''')

import sys
rl = sys.stdin.readline

matrix = [list(map(int, rl().split())) for _ in range(9) ]

max_value = -1
max_row, max_col = -1, -1

for i in range(9):
    for j in range(9):
        if matrix[i][j] > max_value:
            max_value = matrix[i][j]
            max_row = i
            max_col = j


print(max_value)
print(max_row + 1, max_col + 1)
```

```python
# 3	10798	세로읽기

sys.stdin = StringIO('''AABCDD
afzz
09121
a8EWg6
P5h3kx''')

import sys
rl = sys.stdin.readline

data = [['#'] * 15 for _ in range(5)]

for i in range(5):
    str = rl().rstrip()
    for j in range(len(str)):
        data[i][j] = str[j]
result = []
for col in range(15):
    for row in range(5):
        c = data[row][col]
        if not c == '#':
            result.append(data[row][col])

print("".join(result))

```

```python
# 4	2563	색종이

sys.stdin = StringIO('''3
3 7
15 7
5 2''')

import sys
rl = sys.stdin.readline

length = 100
white = [[0] * length for _ in range(length)]

n = int(rl())
for _ in range(n):
    x, y = map(int, rl().split())

    col_start = x
    col_end = x + 9

    row_start = length-1 - y
    row_end = length-1 - y - 9

    for i in range(col_start, col_end + 1):
        for j in range(row_end, row_start + 1):
            white[j][i] = 1

area = 0;
for row in white:
    area += sum(row)
print(area)
```
