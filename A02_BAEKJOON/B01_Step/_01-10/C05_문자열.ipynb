```python
# 6	10809	알파벳 찾기
stdin = StringIO('''baekjoon''')

rl = stdin.readline
word = rl().rstrip()
alphabet = [-1]*26
for i in range(len(word)):
    idx = ord(word[i]) - ord('a') 
    if alphabet[idx] == -1:
        alphabet[idx] = i
print(" ".join(map(str, alphabet)))
```

```python
# 7	2675	문자열 반복

stdin = StringIO('''2
3 ABC
5 /HTP''')

rl = stdin.readline
n = int(rl())
for _ in range(n):
    repeat, word = rl().rstrip().split()
    result = (w * int(repeat) for w in word )
    print("".join(result))
```

```python
# 10	5622	다이얼
stdin = StringIO('''UNUCIC''')

rl = stdin.readline
word = rl().rstrip()

dial_groups = [
    'ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQRS', 'TUV', 'WXYZ'
]
total_time=0

for w in word:
    for group in dial_groups:
        if w in group:
            total_time += dial_groups.index(group) + 3

print(total_time)
```
