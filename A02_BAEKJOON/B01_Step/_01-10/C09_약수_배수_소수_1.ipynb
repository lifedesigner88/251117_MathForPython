```python
# 1	5086	배수와 약수

sys.stdin = StringIO('''8 16
32 4
17 5
0 0''')

import sys
rl = sys.stdin.readline

while True:
    a, b = map(int, rl().split())
    if a == b == 0:
        break
    if a % b == 0:
        print('multiple')
    elif b % a == 0:
        print('factor')
    else:
        print('neither')
```

```python

# 2	2501	약수 구하기
sys.stdin = StringIO('''2735 1''')

import sys
num, n = map(int, sys.stdin.readline().split())

answers = []
for div in range(1, num + 1):
    if num % div == 0:
        if len(answers) == n-1:
            print(div)
            break
        answers.append(div)
else: 
    print('0')




```
