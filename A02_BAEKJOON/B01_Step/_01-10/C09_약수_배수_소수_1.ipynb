```python
# 1	5086	배수와 약수

sys.stdin = StringIO('''8 16
32 4
17 5
0 0''')

import sys
rl = sys.stdin.readline

while True:
    a, b = map(int, rl().split())
    if a == b == 0:
        break
    if a % b == 0:
        print('multiple')
    elif b % a == 0:
        print('factor')
    else:
        print('neither')
```

```python

# 2	2501	약수 구하기
sys.stdin = StringIO('''2735 1''')

import sys
num, n = map(int, sys.stdin.readline().split())

answers = []
for div in range(1, num + 1):
    if num % div == 0:
        if len(answers) == n-1:
            print(div)
            break
        answers.append(div)
else:
    print('0')
```

```python
# 3	9506	약수들의 합

sys.stdin = StringIO('''6
12
28
-1''')

import sys
from math import sqrt
rl = sys.stdin.readline

while True:
    num = int(rl())
    if num == -1:
        break
    
    divs = [1]
    for d in range(2, int(sqrt(num))+1):
        if num % d == 0:
            if d**2 == num:
                divs.append(d)
            else:
                divs.append(d)
                divs.append(num//d)
    if sum(divs) == num:
        divs.sort()
        print(num, "=", " + ".join(map(str, divs)))
    else:
        print(num, "is NOT perfect.")
```

```python
# 4	1978	소수 찾기
sys.stdin = StringIO('''4
1 3 5 7''')

import sys
from math import sqrt
rl = sys.stdin.readline
rl()

nums = list(map(int, rl().split()))
answer = 0
for num in nums:
    if num != 1:
        for div in range(2, int(sqrt(num)) + 1):
            if num%div == 0:
                break
        else:
            answer += 1
print(answer)
```

```python
# 5	2581	소수
sys.stdin = StringIO('''60
100''')

import sys
from math import isqrt

rl = sys.stdin.readline

start = int(rl())
end = int(rl())

prime_flags = [True] * (end + 1)
prime_flags[0] = prime_flags[1] = False

for i in range(2, isqrt(end) + 1):
    if prime_flags[i]:
        for np in range(i*i, end + 1, i):
            prime_flags[np] = False

primes = [num for num in range(start, end + 1) if prime_flags[num]]

if primes:
    print(sum(primes))
    print(primes[0])
else:
    print(-1)
```
