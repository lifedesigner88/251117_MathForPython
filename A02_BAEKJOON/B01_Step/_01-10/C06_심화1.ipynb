```python
# 2	3003	킹, 퀸, 룩, 비숍, 나이트, 폰

sys.stdin = StringIO('''2 1 2 1 2 1''')

import sys
rl = sys.stdin.readline

chess = [1, 1, 2, 2, 2, 8]
exist = map(int, rl().split())
result = (str(a - b) for a, b in zip(chess, exist))

print(" ".join(result))
```

```python
# 3	2444	별 찍기 - 7
sys.stdin = StringIO('''5''')

import sys
n = int(sys.stdin.readline())
length = 2*n - 1
middle = n - 1
for i in range(length):
     blank = abs(i-middle)
     star = length - 2*blank
     print(" " * blank + "*" * star)
```

```python
# 4	10988	팰린드롬인지 확인하기

sys.stdin = StringIO('''level''')

import sys
word = sys.stdin.readline().rstrip()
length = len(word)

result = 1

for i in range(length//2):
    if not word[i] == word[length-1 - i]:
        result = 0
        break

print(result)
```

```python
# 5	1157	단어 공부

sys.stdin = StringIO('''MissisSsipi''')

import sys
word = sys.stdin.readline().rstrip().upper()

word_counter= [0]*26

for w in word:
    index = ord(w)-ord('A')
    word_counter[index] += 1

max_value = max(word_counter)
max_count = word_counter.count(max_value)

if max_count > 1:
    print('?')
else:
    print(chr(ord('A') + word_counter.index(max_value)))
```

```python
# 6	2941	크로아티아 알파벳

sys.stdin = StringIO('''ljes=njak''')

import sys
word = sys.stdin.readline().rstrip()

symbols = ['c=', 'c-', 'dz=', 'd-', 'lj', 'nj', 's=', 'z=']
for sym in symbols:
    word = word.replace(sym, "#")
print(len(word))
```

```python
# 7	1316	그룹 단어 체커
sys.stdin = StringIO('''9
aaa
aaazbz
babb
aazz
azbz
aabbaa
abacc
aba
zzaz''')

import sys
rl = sys.stdin.readline

n = int(rl())
answer = 0
for _ in range(n):
    preword = '@'
    word = rl().rstrip()
    flag = [0]*26
    is_group_num = True
    for w in word:
        index = ord(w)-ord('a')
        if preword == w:
            continue
        elif flag[index] == 1:
            is_group_num = False
            break
        else:
            preword = w
            flag[index] = 1
    if is_group_num:
        answer += 1
print(answer)
```

```python
# 8	25206	너의 평점은

sys.stdin = StringIO('''ObjectOrientedProgramming1 3.0 A+
IntroductiontoComputerEngineering 3.0 A+
ObjectOrientedProgramming2 3.0 A0
CreativeComputerEngineeringDesign 3.0 A+
AssemblyLanguage 3.0 A+
InternetProgramming 3.0 B0
ApplicationProgramminginJava 3.0 A0
SystemProgramming 3.0 B0
OperatingSystem 3.0 B0
WirelessCommunicationsandNetworking 3.0 C+
LogicCircuits 3.0 B0
DataStructure 4.0 A+
MicroprocessorApplication 3.0 B+
EmbeddedSoftware 3.0 C0
ComputerSecurity 3.0 D+
Database 3.0 C+
Algorithm 3.0 B0
CapstoneDesigninCSE 3.0 B+
CompilerDesign 3.0 D0
ProblemSolving 4.0 P''')

import sys

rl = sys.stdin.readline
scores = (
    'A+',	4.5,
    'A0',	4.0,
    'B+',	3.5,
    'B0',	3.0,
    'C+',	2.5,
    'C0',	2.0,
    'D+',	1.5,
    'D0',	1.0,
    'F',	0.0
)

total_degree = 0
total_score = 0

for _ in range(20):
    a, b, c =rl().split()
    degree = float(b)
    if c == 'P':
        continue
    else:
        index = scores.index(c)
        score = scores[index + 1]
        total_degree += degree
        total_score += degree * score

print(total_score/total_degree)     

```
