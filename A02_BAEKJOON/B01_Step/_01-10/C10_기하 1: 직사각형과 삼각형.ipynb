```python
# 3	3009	 네 번째 점
input = StringIO('''5 5
5 7
7 5''').readline

xs, ys = [], []

for _ in range(3):
    a, b = map(int, input().split())
    for val, arr in ((a, xs), (b, ys)):
        arr.remove(val) if val in arr else arr.append(val)

print(xs[0], ys[0])
```

```python
# 5	9063	 대지
input = StringIO('''4
2 1
3 2
5 2
3 4''').readline

n = int(input())

xs, ys = [], []
for _ in range(n):
    x, y = map(int, input().split())
    xs.append(x)
    ys.append(y)
width = max(xs) - min(xs)
heigh = max(ys) - min(ys)
print(width * heigh)

```

```python
# 6	10101	 삼각형 외우기
input = StringIO('''60
70
50''').readline

angles = [int(input()) for _ in range(3)]
angles.sort()
if not sum(angles) == 180:
    print('Error')
elif angles[0] == angles[1] == angles[2]:
    print('Equilateral')
elif angles[0] == angles[1] or angles[1] == angles[2]:
    print('Isosceles')
else:
    print('Scalene')
```

```python
# 7	5073	 삼각형과 세 변
input = StringIO('''7 7 7
6 5 4
3 2 5
6 2 6
0 0 0''').readline

while True:
    a, b, c = map(int, input().split())
    if a==b==c==0:
        break
    
    lst = [a, b, c]
    lst.sort()
    if lst[0] + lst[1] <= lst[2]:
        print('Invalid')
    elif lst[0]==lst[1]==lst[2]:
        print('Equilateral')
    elif lst[0]==lst[1] or lst[1]==lst[2]:
        print('Isosceles')
    else:
        print('Scalene')


```

```python
# 8	14215	 세 막대
input = StringIO('''1 100 1''').readline

a, b, c = map(int, input().split())
sum = a + b + c
longest = max(a, b, c)
two_length = sum - longest

if two_length > longest:
    print(sum)
else:
    print(two_length*2 - 1)

```
