```python
# 1	11279	 최대 힙
import sys

from io import StringIO
sys.stdin = StringIO('''13
0
1
2
0
0
3
2
1
0
0
0
0
0''')

input = sys.stdin.readline

N = int(input())

heap = []
out = []

for _ in range(N):
    num = int(input())
    if num == 0:
        if heap:
            out.append(heap[0])

            if len(heap) > 1:
                last_value = heap.pop()
                heap[0] = last_value
                parent_idx = 0

                while True:
                    left_idx = 2 * parent_idx + 1
                    right_idx = 2 * parent_idx + 2
                    heap_len = len(heap)

                    parent_value = heap[parent_idx]
                    if left_idx >= heap_len:  # 왼쪽 자식 없으면
                        break

                    else:
                        left_value = heap[left_idx] # 왼쪽 자식 값 조회, 

                        if right_idx >= heap_len:  # 왼쪽 값만 존재하고 오른쪽 없으면

                            if left_value <= parent_value: # 왼쪽 값만 부모와 비교 부모가 더 크면 종료. 
                                break

                            else: # 왼쪽 자식이 더 크면 바꾸고 왼쪽 검사. 
                                heap[parent_idx], heap[left_idx] = left_value, parent_value
                                parent_idx = left_idx

                        else:  # 왼쪽 O, 오른쪽 O 둘다 값이 있으면
                            right_value = heap[right_idx]  # 오른쪽 값 조회. 

                            if left_value >= right_value:  # 왼쪽이 오른쪽 보다 크거나 같을 때

                                if left_value <= parent_value:  # 부모가 제일 큼 종료
                                    break

                                else: # 왼쪽이 더 크면 바꾸고 왼쪽 검사. 
                                    heap[parent_idx], heap[left_idx] = left_value, parent_value
                                    parent_idx = left_idx
                            else:

                                if right_value <= parent_value:  # 부모가 제일 큼 종료
                                    break

                                else: # 오른쪽이 더 크면 바꾸고 오른쪽 재검사. 
                                    heap[parent_idx], heap[right_idx] = right_value, parent_value
                                    parent_idx = right_idx

            else:
                heap.pop()  # empty

        else:
            out.append(0)
    else:
        # push
        last_idx = len(heap)
        heap.append(num)
        if last_idx != 0:
            child_idx = last_idx
            while child_idx > 0:
                parent_idx = (child_idx - 1) // 2
                c_value = heap[child_idx]
                p_value = heap[parent_idx]
                if p_value >= c_value:
                    break
                else:
                    heap[parent_idx], heap[child_idx] = c_value, p_value
                    child_idx = parent_idx

print("\n".join(map(str, out)))
```

```python
# 2	1927	 최소 힙
import sys

from io import StringIO
sys.stdin = StringIO('''9
0
12345678
1
2
0
0
0
0
32''')

input = sys.stdin.readline

N = int(input())

out = list()
heap = list()

for _ in range(N):
    num = int(input())

    if num == 0: # pop()
        if heap: # 1개 이상
            out.append(heap[0]) 
            temp = heap.pop()

            if heap: # 1개 초과 (2개 이상) (pop이후 남아 있는경우)
                heap[0] = temp
                parent_idx = 0
                while True:
                    left_idx = 2*parent_idx + 1
                    right_idx = 2*parent_idx + 2
                    target_idx = parent_idx

                    if left_idx < len(heap) and heap[left_idx] < heap[target_idx]:
                        target_idx = left_idx
                    if right_idx < len(heap) and heap[right_idx] < heap[target_idx]:
                        target_idx = right_idx

                    if target_idx == parent_idx:
                        break
                    
                    heap[target_idx], heap[parent_idx] = heap[parent_idx], heap[target_idx]
                    parent_idx = target_idx

            continue
        out.append(0) # 0개 일때

    else: # push()
        heap.append(num)
        child_idx = len(heap) - 1
        while child_idx > 0:
            parent_idx = (child_idx - 1) // 2
            if heap[child_idx] >= heap[parent_idx]:
                break
            heap[child_idx], heap[parent_idx] = heap[parent_idx], heap[child_idx]
            child_idx = parent_idx
    
print("\n".join(map(str, out)))
```

```python
# 3	11286	 절댓값 힙

import sys
from heapq import heappop, heappush

from io import StringIO
sys.stdin = StringIO('''18
1
-1
0
0
0
1
1
-1
-1
2
-2
0
0
0
0
0
0
0''')

input = sys.stdin.readline
N = int(input())

heap = []
out = []

for _ in range(N):
    
    num = int(input())
    
    if num == 0:
        if not heap:
            out.append(0)
            continue
        out.append(heappop(heap)[1])
    else:
        heappush(heap, (abs(num), num))

print("\n".join(map(str, out)))
```

```python
# 4	2075	 N번째 큰 수
import sys
from heapq import heapify, heappushpop

from io import StringIO
sys.stdin = StringIO('''5
12 7 9 15 5
13 8 11 19 6
21 10 26 31 16
48 14 28 35 25
52 20 32 41 49''')
input = sys.stdin.readline

N = int(input())

heap = list(map(int, input().split()))
heapify(heap)

for _ in range(N - 1):
    for x in map(int, input().split()):
        heappushpop(heap, x)

print(heap[0])
```

```python
# 5	2696	 중앙값 구하기
import sys
from heapq import heappush,heappushpop

from io import StringIO
sys.stdin = StringIO('''3
9
1 2 3 4 5 6 7 8 9
9
9 8 7 6 5 4 3 2 1
23
23 41 13 22 -3 24 -31 -11 -8 -7
3 5 103 211 -311 -45 -67 -73 -81 -99
-33 24 56''')

input = sys.stdin.readline

N = int(input())

for _ in range(N):
    M = int(input()) # 홀수 

    max_heap = [] # 중앙 값 보유.
    min_heap = [] 
    out = [] # 결과.

    for _ in range(M//10 + 1):
        for i, num in enumerate(map(int, input().split())):
            if not max_heap: # 시작
                max_heap.append(-num)
                out.append(num)
                continue
            if i & 1: # 값이 있고, 짝수번째 입력 (인덱스는 홀 수 1, 3, 5) (맥스 길이가  min길이보다 1개 더 긴 상태.)
                heappush(min_heap, -heappushpop(max_heap, -num))

            else: # (홀수번째 입력) 짝수 인덱스 이면(0, 2, 4) .(max, min 길이가 같은 상태)
                center = -max_heap[0]
                if center >= num:
                    heappush(max_heap, -num)
                else:
                    heappush(max_heap, -heappushpop(min_heap, num))
                out.append(-max_heap[0])

    out_length = len(out)
    print(out_length)

    if out_length <= 10:
        print(*out)
    else:
        for i in range(out_length//10 + 1):
            print(*out[0 + i*10 : 10 + i*10])
```

```python
# 6	1202	 보석 도둑
import sys
from heapq import heappush, heappop

from io import StringIO
sys.stdin= StringIO('''2 1
5 10
100 100
11''')


input = sys.stdin.readline

N, K = map(int, input().split()) # 보석개수, 가방개수

gram_price = [tuple(map(int, input().split())) for _ in range(N)]
gram_price.sort(key=lambda x: (x[0]))

bag_mass = [int(input()) for _ in range(K)]
bag_mass.sort()


gram_price_idx = 0
prices = 0
jewels = []
for gram in bag_mass:
    for i in range(gram_price_idx, len(gram_price)):
        jewel = gram_price[i]
        if jewel[0] <= gram:
            heappush(jewels, -jewel[1])
        else:
            gram_price_idx = i
            break
    else: 
        gram_price_idx = N
    
    if jewels:
        prices += -heappop(jewels)

print(prices)
```
