```python
# 1	11279	 최대 힙
import sys

from io import StringIO
sys.stdin = StringIO('''13
0
1
2
0
0
3
2
1
0
0
0
0
0''')

input = sys.stdin.readline

N = int(input())

heap = []
out = []

for _ in range(N):
    num = int(input())
    if num == 0:
        if heap:
            out.append(heap[0])

            if len(heap) > 1:
                last_value = heap.pop()
                heap[0] = last_value
                parent_idx = 0

                while True:
                    left_idx = 2 * parent_idx + 1
                    right_idx = 2 * parent_idx + 2
                    heap_len = len(heap)

                    parent_value = heap[parent_idx]
                    if left_idx >= heap_len:  # 왼쪽 자식 없으면
                        break

                    else:
                        left_value = heap[left_idx] # 왼쪽 자식 값 조회, 

                        if right_idx >= heap_len:  # 왼쪽 값만 존재하고 오른쪽 없으면

                            if left_value <= parent_value: # 왼쪽 값만 부모와 비교 부모가 더 크면 종료. 
                                break

                            else: # 왼쪽 자식이 더 크면 바꾸고 왼쪽 검사. 
                                heap[parent_idx], heap[left_idx] = left_value, parent_value
                                parent_idx = left_idx

                        else:  # 왼쪽 O, 오른쪽 O 둘다 값이 있으면
                            right_value = heap[right_idx]  # 오른쪽 값 조회. 

                            if left_value >= right_value:  # 왼쪽이 오른쪽 보다 크거나 같을 때

                                if left_value <= parent_value:  # 부모가 제일 큼 종료
                                    break

                                else: # 왼쪽이 더 크면 바꾸고 왼쪽 검사. 
                                    heap[parent_idx], heap[left_idx] = left_value, parent_value
                                    parent_idx = left_idx
                            else:

                                if right_value <= parent_value:  # 부모가 제일 큼 종료
                                    break

                                else: # 오른쪽이 더 크면 바꾸고 오른쪽 재검사. 
                                    heap[parent_idx], heap[right_idx] = right_value, parent_value
                                    parent_idx = right_idx

            else:
                heap.pop()  # empty

        else:
            out.append(0)
    else:
        # push
        last_idx = len(heap)
        heap.append(num)
        if last_idx != 0:
            child_idx = last_idx
            while child_idx > 0:
                parent_idx = (child_idx - 1) // 2
                c_value = heap[child_idx]
                p_value = heap[parent_idx]
                if p_value >= c_value:
                    break
                else:
                    heap[parent_idx], heap[child_idx] = c_value, p_value
                    child_idx = parent_idx

print("\n".join(map(str, out)))
```

```python
# 1	11279	 최대 힙 최적화. 

import sys
input = sys.stdin.readline

N = int(input())
heap = []
out = []

for _ in range(N):
    x = int(input())

    # pop
    if x == 0:
        if not heap:
            out.append(0)
            continue

        out.append(heap[0])

        last = heap.pop()
        if heap:
            heap[0] = last
            idx = 0
            size = len(heap)

            while True:
                left = 2 * idx + 1
                right = 2 * idx + 2
                largest = idx

                if left < size and heap[left] > heap[largest]:
                    largest = left
                if right < size and heap[right] > heap[largest]:
                    largest = right

                if largest == idx:
                    break

                heap[idx], heap[largest] = heap[largest], heap[idx]
                idx = largest

    # push
    else:
        heap.append(x)
        idx = len(heap) - 1

        while idx > 0:
            parent = (idx - 1) // 2
            if heap[parent] >= heap[idx]:
                break
            heap[parent], heap[idx] = heap[idx], heap[parent]
            idx = parent

print("\n".join(map(str, out)))
```
