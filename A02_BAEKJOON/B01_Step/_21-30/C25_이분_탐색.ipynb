```python
# 1	1920	 수 찾기
import sys

from io import StringIO
sys.stdin = StringIO('''5
4 1 5 2 3
5
1 3 7 9 5''')

input = sys.stdin.readline

N = int(input())
nums = list(map(int, input().split()))

M = int(input())
targets = list(map(int, input().split()))

nums.sort()

out = []
for goal in targets:
    
    start, end = 0, N - 1
    found = 0

    while start <= end:
        mid = (end + start) >> 1
        num = nums[mid]
        if num == goal:
            found = 1
            break
        elif num < goal:
            start = mid + 1
        else:
            end = mid - 1

    out.append(str(found))

print("\n".join(out))

# (Set 이용)
# n = int(input())
# n_set = set(map(int, input().split()))

# m = int(input())
# m_arr = map(int, input().split())
# print("\n".join("1" if x in n_set else "0" for x in m_arr)) 
```

```python
# 2	10816	 숫자 카드 2
import sys 

from io import StringIO
sys.stdin = StringIO('''10
6 3 2 10 10 10 -10 -10 7 3
8
10 9 -5 2 3 4 5 -10''')


input = sys.stdin.readline

def lower_bound(arr, x):
    start, end = 0, len(arr)

    while start < end:
        mid = start + end >> 1
        if arr[mid] >= x:
            end = mid
        else:
            start = mid + 1
    return start

def upper_bound(arr, x):
    start, end = 0, len(arr)
    while start < end:
        mid = start + end >> 1
        if arr[mid] > x:
            end = mid
        else:
            start = mid + 1
    return start


N = int(input())
nums = list(map(int, input().split()))
nums.sort()

M = int(input())
goals = list(map(int, input().split()))

print(*[str(upper_bound(nums, goal) - lower_bound(nums, goal)) for goal in goals])
```

```python
# 3	1654	 랜선 자르기
import sys 

from io import StringIO
sys.stdin = StringIO('''4 11
802
743
457
539''')


input = sys.stdin.readline

K, N = map(int, input().split())
def how_many(arr, x):
    count = 0
    for line in arr:
        count += line // x
    return count

lines = [int(input()) for _ in range(K)]

start = 1
end = max(lines) + 1

while start < end:
    mid = (start + end + 1) // 2
    goal = N
    if how_many(lines, mid) >= goal:
        start = mid
    else:
        end = mid - 1
    
print(start)
```

```python
# 4	2805	 나무 자르기
import sys

from io import StringIO
sys.stdin = StringIO('''5 20
4 42 40 26 46''')

input = sys.stdin.readline
N, M = map(int, input().split())
trees = list(map(int, input().split()))


def cut_tree(length: int) -> int:
    cuted_tree = 0
    for tree in trees:
        if tree > length:
            cuted_tree += tree - length
            if cuted_tree >= M:
                return cuted_tree
    return cuted_tree

start = 0
end = max(trees)

while start < end:
    mid = (start + end + 1) // 2
    mid_value = cut_tree(mid)
    if mid_value < M:
        end = mid - 1
    else:
        start = mid

print(start)
```

```python
import sys
import bisect

from io import StringIO
sys.stdin = StringIO('''5 20
4 42 40 26 46''')

input = sys.stdin.readline

N, M = map(int, input().split())
trees = sorted(map(int, input().split()))

# prefix[i] = trees[0] + ... + trees[i-1]
prefix = [0] * (N + 1)
for i in range(N):
    prefix[i + 1] = prefix[i] + trees[i]

def cut_amount(h: int) -> int:
    # h보다 큰 나무의 첫 인덱스
    idx = bisect.bisect_right(trees, h)

    cnt = N - idx
    if cnt == 0:
        return 0

    total_height = prefix[N] - prefix[idx]
    return total_height - h * cnt

lo, hi = 0, trees[-1]
while lo < hi:
    mid = (lo + hi + 1) // 2
    if cut_amount(mid) >= M:
        lo = mid
    else:
        hi = mid - 1

print(lo)
```

```python
# 5	2110	 공유기 설치
import sys

from io import StringIO
sys.stdin = StringIO('''5 3
1
2
8
4
9''')

input = sys.stdin.readline

N, C = map(int, input().split())
houses = [int(input()) for _ in range(N)]
houses.sort()

start = 1
end = houses[-1] - houses[0]
answer = 0 # 최소 길이 값을 정하고 C 갯수 만큼 설치 가능한지 체크, 가능하면 길이를 늘리고, 불가능 하면 줄여서 탐색. 

while start <= end:
    mid = (start + end) // 2
    count = 1
    last = houses[0]
    for house in houses[1:]:
        if house - last >= mid:
            count += 1
            last = house
    if count >= C:
        answer = mid
        start = mid + 1
    else:
        end = mid - 1

print(answer)


```

```python
# 6	1300	 K번째 수
import sys

from io import StringIO
sys.stdin = StringIO('''3
7''')

input = sys.stdin.readline

N = int(input())
K = int(input())


def count(x):
    smalls_same_then_x = 0

    for i in range(1, N + 1):
        if min(N, x//i) == 0:
            break
        smalls_same_then_x += min(N, x//i)
    return smalls_same_then_x


start = 1
end = N**2 + 1

while start < end:
    mid = (start + end) // 2
    mid_value = count(mid)
    if mid_value >= K:
        end = mid
    else:
        start = mid + 1

print(start)
```
