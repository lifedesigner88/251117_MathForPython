```python
# 4	9370	 미확인 도착지
import sys

from io import StringIO
sys.stdin = StringIO('''2
5 4 2
1 2 3
1 2 6
2 3 2
3 4 4
3 5 3
5
4
6 9 2
2 3 1
1 2 1
1 3 3
2 4 4
2 5 5
3 4 3
3 6 2
4 5 4
4 6 3
5 6 7
5
6''')


input = sys.stdin.readline

from heapq import heappush, heappop

T = int(input())
INF = sys.maxsize

for _ in range(T):
    n, m, t = map(int, input().split())  # n 교차(노드) 개수, 도로개수(연결), 목적지 후보의 개수
    start, g, h = map(int, input().split())  # 출발지, g -> h 지나감. ( 필수로 지나감 ) 

    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        a, b, w = map(int, input().split())
        graph[a].append((b, w))
        graph[b].append((a, w))
    
    candidates = []
    for _ in range(t):
        candidates.append(int(input()))

    # 출발지 지정 후 다익스라 리턴 함수 구현

    def dextra(s):
        min_dist = [INF] * (n + 1)
        min_dist[s] = 0
        pq = [(0, s)] # 시작노드. 

        while pq:
            cur_dist, cur = heappop(pq)

            if cur_dist > min_dist[cur]:
                continue

            for nxt, cost in graph[cur]:
                nxt_dist = cur_dist + cost
                if nxt_dist < min_dist[nxt]:
                    min_dist[nxt] = nxt_dist
                    heappush(pq, (nxt_dist, nxt))
                    
        return min_dist

    # start 로 시작해서 최단거리 다익스트라 구함. 

    from_s = dextra(start)
    from_g = dextra(g)
    from_h = dextra(h)


    # 지정 노드를 반드시 지나가면서 최단거리 다 익스라 구함. 
    # 두개 비교해서 일치하는 값만 리턴. 

    # s -> g -> h -> c
    # s -> h -> g -> c

    def check(c):
        path1 = from_s[g] + from_g[h] + from_h[c]
        path2 = from_s[h] + from_h[g] + from_g[c]

        return from_s[c] == min(path1, path2) 

    possible = [candi for candi in candidates if check(candi)]
    possible.sort()
    print(*possible)

```

```python
# 4	9370	 미확인 도착지 (홀짝 트릭)
import sys

from io import StringIO
sys.stdin = StringIO('''2
5 4 2
1 2 3
1 2 6
2 3 2
3 4 4
3 5 3
5
4
6 9 2
2 3 1
1 2 1
1 3 3
2 4 4
2 5 5
3 4 3
3 6 2
4 5 4
4 6 3
5 6 7
5
6''')


input = sys.stdin.readline

from heapq import heappush, heappop

T = int(input())
INF = sys.maxsize

for _ in range(T):
    n, m, t = map(int, input().split())  # n 교차(노드) 개수, 도로개수(연결), 목적지 후보의 개수
    start, g, h = map(int, input().split())  # 출발지, g -> h 지나감. ( 필수로 지나감 ) 

    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        a, b, w = map(int, input().split())
        double = w * 2

        if (a, b) == (g, h) or (a, b) == (h, g):
            double -= 1

        graph[a].append((b, double))
        graph[b].append((a, double))
    
    candidates = []
    for _ in range(t):
        candidates.append(int(input()))

    # 출발지 지정 후 다익스라 리턴 함수 구현

    def dextra(s):
        min_dist = [INF] * (n + 1)
        min_dist[s] = 0
        pq = [(0, s)] # 시작노드. 

        while pq:
            cur_dist, cur = heappop(pq)

            if cur_dist > min_dist[cur]:
                continue

            for nxt, cost in graph[cur]:
                nxt_dist = cur_dist + cost
                if nxt_dist < min_dist[nxt]:
                    min_dist[nxt] = nxt_dist
                    heappush(pq, (nxt_dist, nxt))
                    
        return min_dist

    # start 로 시작해서 최단거리 다익스트라 구함. 

    from_s = dextra(start)
    possible = [c for c in candidates if from_s[c] != INF and from_s[c] % 2 == 1]
    possible.sort()
    print(*possible)

```

```python
# 7	1956	 운동 
import sys

from io import StringIO
sys.stdin = StringIO('''3 4
1 2 1
3 2 1
1 3 5
2 3 2''')

input = sys.stdin.readline

V, E = map(int, input().split())

INF = sys.maxsize

graph = [[] for _ in range(V + 1)]

for _ in range(E):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))

from heapq import heappop, heappush


def dijkstra(start):
    min_dist = [INF] * (V + 1)
    min_dist[start] = 0

    pq = [(0, start)]

    while pq:
        cur_dist, cur = heappop(pq)

        if cur_dist <= min_dist[cur]:
            
            for nxt, cost in graph[cur]:
                nxt_dist = cur_dist + cost
                if nxt_dist < min_dist[nxt]:
                    min_dist[nxt] = nxt_dist
                    heappush(pq, (nxt_dist, nxt))

    return min_dist

answer = INF

for i in range(1, V + 1):
    dist = dijkstra(i)
    for u in range(1, V + 1):
        for v, w in graph[u]:
            if v == i and dist[u] != INF:
                answer = min(answer, dist[u] + w)

print(answer if answer != INF else -1)
```
