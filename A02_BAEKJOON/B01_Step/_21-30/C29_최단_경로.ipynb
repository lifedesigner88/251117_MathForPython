```python
# 1	1753	 최단경로

import sys

from io import StringIO
sys.stdin = StringIO('''5 6
1
5 1 1
1 2 2
1 3 3
2 3 4
2 4 5
3 4 6''')

input = sys.stdin.readline

V, E = map(int, input().split())

start = int(input())

graph = [[] for _ in range(V + 1)]

INF = sys.maxsize
dist = [INF] * (V + 1)
dist[start] = 0

for _ in range(E):
    u, v, w, = map(int, input().split())
    graph[u].append((v, w))

from heapq import heappush, heappop

pq = []
heappush(pq, (0, start))

while pq:
    cur_dist, cur = heappop(pq)

    if cur_dist > dist[cur]:
        continue
    
    for nxt, cost in graph[cur]:
        new_dist = cur_dist + cost
        if new_dist < dist[nxt]:
            dist[nxt] = new_dist
            heappush(pq, (new_dist, nxt))

for i in range(1, V + 1):
    if dist[i] == INF:
        print("INF")
    else:
        print(dist[i])

```

```python
# 2	1504	 특정한 최단 경로

from heapq import heappop, heappush
import sys

from io import StringIO
sys.stdin = StringIO('''4 6
1 2 3
2 3 3
3 4 1
1 3 5
2 4 5
1 4 4
2 3''')

input = sys.stdin.readline

N, E = map(int, input().split())

INF = sys.maxsize

graph = [[] for _ in range(N + 1)]
for _ in range(E):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))
    graph[b].append((a, c))


def dextra(start):

    dist = [INF] * (N + 1)
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        cur_dist, cur = heappop(pq)
        if cur_dist <= dist[cur]:
            for nxt, cost in graph[cur]:
                new_dist = cur_dist + cost
                if new_dist < dist[nxt]:
                    dist[nxt] = new_dist
                    heappush(pq, (new_dist, nxt))

    return dist

v1, v2 = map(int, input().split())

start = dextra(1)
# 1 -> v1, 1-> v2

# v1 -> v2, v1-> N
first = dextra(v1)

# v2 -> v1, V2 -> N
second = dextra(v2)

# 1-> v1 -> v2 -> N
# 1-> v2 -> v1 -> N
path1 = start[v1] + first[v2] + second[N]
path2 = start[v2] + second[v1] + first[N]

ans = min(path1, path2)

print(ans if ans < INF else "-1")
```
