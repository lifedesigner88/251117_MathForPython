```python
# 1	11659	 구간 합 구하기 4

import sys

from io import StringIO
sys.stdin = StringIO('''5 3
5 4 3 2 1
1 3
2 4
5 5''')

input = sys.stdin.readline

N, M = map(int, input().split())
nums = list(map(int, input().split()))

# prefix[k] = nums[0] ~ nums[k-1] 합 (prefix[0]=0)
prefix = [0] * (N + 1)
for i in range(1, N + 1):
    prefix[i] = prefix[i - 1] + nums[i - 1]

out = []
for _ in range(M):
    i, j = map(int, input().split())
    out.append(str(prefix[j] - prefix[i - 1]))

print("\n".join(out))
```

```python
# 2	2559	 수열

import sys

from io import StringIO
sys.stdin = StringIO('''10 2
3 -2 -4 -9 0 3 7 13 8 -3''')

input = sys.stdin.readline

N, K = map(int, input().split())
nums = list(map(int, input().split()))

window_sum  = sum(nums[0:K])
best = sum(nums[0:K])

for i in range(K, N):
    window_sum  += nums[i-K] + nums[i]
    if  best < window_sum:
        best = window_sum

print(best)
```

```python
# 3	16139	 인간-컴퓨터 상호작용
import sys

from io import StringIO
sys.stdin = StringIO('''seungjaehwang
4
a 0 5
a 0 6
a 6 10
a 7 10''')
input = sys.stdin.readline

S = input().strip()
q = int(input())
n = len(S)

# prefix[c][i] = S[:i]에서 문자 c의 개수 (i는 0..n)
prefix = [[0] * (n + 1) for _ in range(26)]

for i, ch in enumerate(S, start=1):
    idx = ord(ch) - 97
    for c in range(26):
        prefix[c][i] = prefix[c][i - 1]
    prefix[idx][i] += 1

out = []
for _ in range(q):
    a, l, r = input().split()
    l = int(l); r = int(r)
    c = ord(a) - 97
    out.append(str(prefix[c][r + 1] - prefix[c][l]))

print("\n".join(out))
```

```python
# 3	16139	 인간-컴퓨터 상호작용
import sys

from io import StringIO
sys.stdin = StringIO('''seungjaehwang
4
a 0 5
a 0 6
a 6 10
a 7 10''')
input = sys.stdin.readline

S = input().strip()
q = int(input())
n = len(S)

# prefix[i][c] = S[:i]에서 문자 c의 개수
prefix = [[0]*26 for _ in range(n+1)]

for i, ch in enumerate(S, 1):
    row = prefix[i]
    prev = prefix[i-1]
    row[:] = prev            # C레벨로 26개 복사 (파이썬 for 26번보다 보통 빠름)
    row[ord(ch)-97] += 1

out = []
for _ in range(q):
    a, l, r = input().split()
    c = ord(a) - 97
    l = int(l); r = int(r) + 1
    out.append(str(prefix[r][c] - prefix[l][c]))

sys.stdout.write("\n".join(out))
```

```python
# 4	10986	 나머지 합
import sys

from io import StringIO
sys.stdin = StringIO('''5 3
1 2 3 1 2''')

input = sys.stdin.readline

N, M = map(int, input().split())
A = tuple(map(int, input().split()))

prefix = [0] * (N + 1)
counter = 0

for i in range(1, N + 1):
    prefix[i] = (prefix[i - 1] + A[i - 1]) % M
    if prefix[i] == 0:          # 0..i-1 구간
        counter += 1

for i in range(1, N):           # i=1부터: 0..k 구간 중복 방지
    for j in range(i + 1, N + 1):
        if (prefix[j] - prefix[i]) % M == 0:
            counter += 1

print(counter)
```

```python
import sys

from io import StringIO
sys.stdin = StringIO('''5 3
1 2 3 1 2''')

input = sys.stdin.readline

N, M = map(int, input().split())
A = map(int, input().split())

cnt = [0] * M
prefix = 0

# prefix % M 값들의 빈도를 센다
# prefix가 0인 케이스를 조합으로 자연스럽게 처리하려고 cnt[0]=1로 시작
cnt[0] = 1

for x in A:
    prefix = (prefix + x) % M
    cnt[prefix] += 1

# 같은 나머지에서 2개를 고르는 조합 수를 합산
ans = 0
for c in cnt:
    ans += c * (c - 1) // 2

print(ans)
```

```python
# 5	11660	 구간 합 구하기 5
import sys

from io import StringIO
sys.stdin = StringIO('''4 3
1 2 3 4
2 3 4 5
3 4 5 6
4 5 6 7
2 2 3 4
3 4 3 4
1 1 4 4''')


input = sys.stdin.readline

N, M = map(int, input().split())

matrix = [tuple(map(int, input().split())) for _ in range(N)]


prefix = [[0] * (N + 1) for _ in range(N + 1)]

for i in range(1, N + 1):
    for j in range(1, N + 1):
        prefix[i][j] = (
            prefix[i][j-1] 
            + prefix[i-1][j] 
            - prefix[i-1][j-1] 
            + matrix[i-1][j-1]
        )

out = [0] * M
for i in range(M):
    x1, y1, x2, y2 = map(int, input().split())
    ans = (
        prefix[x2][y2] 
        - prefix[x1-1][y2] 
        - prefix[x2][y1-1] 
        + prefix[x1-1][y1-1]
    )
    out[i] = str(ans) 

print("\n".join(out))

```

```python
# 6	25682	 체스판 다시 칠하기 2
import sys

from io import StringIO
sys.stdin = StringIO('''10 13 10
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
WWWWWWWWWWBWB
WWWWWWWWWWBWB''')


input = sys.stdin.readline

N, M, K = map(int, input().split())


board = [input().strip() for _ in range(N)]

# 2D prefix sum 만들기 함수
def build_prefix(mismatch):
    ps = [[0] * (M + 1) for _ in range(N + 1)]
    for i in range(1, N + 1):
        row_sum = 0
        for j in range(1, M + 1):
            row_sum += mismatch[i - 1][j - 1]
            ps[i][j] = ps[i - 1][j] + row_sum
    return ps

# (i, j)가 시작색 start('B' or 'W')인 체스판과 맞는지 체크
# (i+j)%2==0 이면 start, 아니면 반대색
def expected_color(i, j, start):
    if (i + j) % 2 == 0:
        return start
    return 'W' if start == 'B' else 'B'

# mismatch_B[i][j] = 시작이 'B'인 패턴과 다르면 1
# mismatch_W[i][j] = 시작이 'W'인 패턴과 다르면 1
mismatch_B = [[0] * M for _ in range(N)]
mismatch_W = [[0] * M for _ in range(N)]

for i in range(N):
    row = board[i]
    for j in range(M):
        c = row[j]
        if c != expected_color(i, j, 'B'):
            mismatch_B[i][j] = 1
        if c != expected_color(i, j, 'W'):
            mismatch_W[i][j] = 1

psB = build_prefix(mismatch_B)
psW = build_prefix(mismatch_W)

def rect_sum(ps, x1, y1, x2, y2):
    # (x1,y1)~(x2,y2) inclusive, 0-indexed
    return ps[x2 + 1][y2 + 1] - ps[x1][y2 + 1] - ps[x2 + 1][y1] + ps[x1][y1]

ans = 10**18
for x in range(N - K + 1):
    for y in range(M - K + 1):
        x2, y2 = x + K - 1, y + K - 1
        repaintB = rect_sum(psB, x, y, x2, y2)
        repaintW = rect_sum(psW, x, y, x2, y2)
        ans = min(ans, repaintB, repaintW)

print(ans)

```

```python
# 6	25682	 체스판 다시 칠하기 2
import sys

from io import StringIO
sys.stdin = StringIO('''10 13 10
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
BBBBBBBBWBWBW
BBBBBBBBBWBWB
WWWWWWWWWWBWB
WWWWWWWWWWBWB''')


input = sys.stdin.readline

N, M, K = map(int, input().split())
board = [input().strip() for _ in range(N)]  # bytes

# ps[i][j] = (0,0)~(i-1,j-1)에서 "B로 시작하는 체스패턴"과 불일치 개수
ps = [[0] * (M + 1) for _ in range(N + 1)]

B = ord('B')
W = ord('W')

for i in range(N):
    row = board[i]
    row_sum = 0

    # (i,0)에서 기대 색 (B-start pattern 기준)
    # i가 짝수면 (i,0)은 'B', 홀수면 'W'
    exp = B if (i & 1) == 0 else W

    for j in range(M):
        # 불일치면 1
        if row[j] != exp:
            row_sum += 1

        # prefix 누적
        ps[i + 1][j + 1] = ps[i][j + 1] + row_sum

        # 다음 칸 기대 색 토글
        exp = W if exp == B else B

kk = K * K
ans = 10**18

# KxK 윈도우 합을 O(1)로 계산
for x in range(N - K + 1):
    x2 = x + K
    psx = ps[x]
    psx2 = ps[x2]
    for y in range(M - K + 1):
        y2 = y + K
        repaintB = psx2[y2] - psx[y2] - psx2[y] + psx[y]
        # W-start는 반대패턴이라 kk - repaintB
        if repaintB > kk - repaintB:
            repaintB = kk - repaintB
        if repaintB < ans:
            ans = repaintB

print(ans)

```
