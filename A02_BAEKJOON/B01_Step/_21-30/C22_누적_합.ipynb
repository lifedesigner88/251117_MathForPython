```python
# 1	11659	 구간 합 구하기 4

import sys

from io import StringIO
sys.stdin = StringIO('''5 3
5 4 3 2 1
1 3
2 4
5 5''')

input = sys.stdin.readline

N, M = map(int, input().split())
nums = list(map(int, input().split()))

# prefix[k] = nums[0] ~ nums[k-1] 합 (prefix[0]=0)
prefix = [0] * (N + 1)
for i in range(1, N + 1):
    prefix[i] = prefix[i - 1] + nums[i - 1]

out = []
for _ in range(M):
    i, j = map(int, input().split())
    out.append(str(prefix[j] - prefix[i - 1]))

print("\n".join(out))
```

```python
# 2	2559	 수열

import sys

from io import StringIO
sys.stdin = StringIO('''10 2
3 -2 -4 -9 0 3 7 13 8 -3''')

input = sys.stdin.readline

N, K = map(int, input().split())
nums = list(map(int, input().split()))

window_sum  = sum(nums[0:K])
best = sum(nums[0:K])

for i in range(K, N):
    window_sum  += nums[i-K] + nums[i]
    if  best < window_sum:
        best = window_sum

print(best)
```
