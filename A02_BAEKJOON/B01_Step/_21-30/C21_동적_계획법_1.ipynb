```python
# 1	24416	 알고리즘 수업 - 피보나치 수 1
import sys

from io import StringIO
sys.stdin = StringIO('''5''')

input = sys.stdin.readline
N = int(input())

fib_count = 0
fibonacci_count = 0


def fib(idx):
    global fib_count

    if idx == 1 or idx == 2:
        fib_count += 1
        return 1

    return fib(idx - 1) + fib(idx - 2)


def fibonacci(n):
    global fibonacci_count

    fibos = [1] * (n + 1)
    for i in range(3, n + 1):
        fibonacci_count += 1
        fibos[i] = fibos[i - 1] + fibos[i - 2]
    return fibos[n]

print(fibonacci(N), fibonacci_count)
```
