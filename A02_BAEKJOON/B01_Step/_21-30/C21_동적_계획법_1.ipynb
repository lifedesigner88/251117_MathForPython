```python
# 1	24416	 알고리즘 수업 - 피보나치 수 1
import sys

from io import StringIO
sys.stdin = StringIO('''5''')

input = sys.stdin.readline
N = int(input())

fib_count = 0
fibonacci_count = 0


def fib(idx):
    global fib_count

    if idx == 1 or idx == 2:
        fib_count += 1
        return 1

    return fib(idx - 1) + fib(idx - 2)


def fibonacci(n):
    global fibonacci_count

    fibos = [1] * (n + 1)
    for i in range(3, n + 1):
        fibonacci_count += 1
        fibos[i] = fibos[i - 1] + fibos[i - 2]
    return fibos[n]

print(fibonacci(N), fibonacci_count)
```

```python
# 2	9184	 신나는 함수 실행
import sys

from io import StringIO
sys.stdin = StringIO('''1 1 1
2 2 2
10 4 6
50 50 50
-1 7 18
-1 -1 -1''')

input = sys.stdin.readline

max_val = 21
maps = [[[0]*(max_val) for _ in range(max_val)] for _ in range(max_val)]

for ai in range(0, max_val):
    for bi in range(0, max_val):
        for ci in range(0, max_val):
            if ai == 0 or bi == 0 or ci == 0:
                maps[ai][bi][ci] = 1
            
            elif ai < bi < ci:
                maps[ai][bi][ci] = maps[ai][bi][ci-1] + \
                    maps[ai][bi-1][ci-1] - maps[ai][bi-1][ci]
            else:
                maps[ai][bi][ci] = maps[ai-1][bi][ci] + maps[ai-1][bi - 1][ci] + \
                    maps[ai-1][bi][ci-1] - maps[ai-1][bi-1][ci-1]

def wdp(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return 1 
    if a > 20 or b > 20 or c > 20:
        return maps[20][20][20]
    
    return maps[a][b][c]

while True:
    a, b, c = map(int, input().split())
    if a == b == c == -1:
        break
    print(f"w({a}, {b}, {c}) = {wdp(a, b, c)}")
```

```python
# 2	9184	 신나는 함수 실행

import sys
from functools import lru_cache

from io import StringIO
sys.stdin = StringIO('''1 1 1
2 2 2
10 4 6
50 50 50
-1 7 18
-1 -1 -1''')

input = sys.stdin.readline

@lru_cache(maxsize=None)
def w(a: int, b: int, c: int) -> int:
    if a <= 0 or b <= 0 or c <= 0:
        return 1
    if a > 20 or b > 20 or c > 20:
        return w(20, 20, 20)

    if a < b < c:
        return w(a, b, c-1) + w(a, b-1, c-1) - w(a, b-1, c)
    return w(a-1, b, c) + w(a-1, b-1, c) + w(a-1, b, c-1) - w(a-1, b-1, c-1)

while True:
    a, b, c = map(int, input().split())
    if a == b == c == -1:
        break
    print(f"w({a}, {b}, {c}) = {w(a, b, c)}")
```

```python
# 6	1149	 RGB거리
import sys

from io import StringIO
sys.stdin = StringIO('''8
71 39 44
32 83 55
51 37 63
89 29 100
83 58 11
65 13 15
47 25 29
60 66 19''')

input = sys.stdin.readline

N = int(input())


MAX = 10**7
price = 0
min_price = MAX
colors = ["r", "g", "b"]

prices = [tuple(map(int, input().split())) for _ in range(N)] 
def dfs(house:int, color="s"):
    global min_price, price
    
    if price > min_price:
        return
    
    if house > N :
        min_price = min(min_price, price)
        return 

    for i in range(3):
        if not colors[i] == color:
            price += prices[house-1][i]
            dfs(house + 1, colors[i])
            price -= prices[house-1][i]
dfs(1)
print(min_price)

```

```python
# 6	1149	 RGB거리

import sys

from io import StringIO
sys.stdin = StringIO('''8
71 39 44
32 83 55
51 37 63
89 29 100
83 58 11
65 13 15
47 25 29
60 66 19''')

sys.setrecursionlimit(10**6)
input = sys.stdin.readline

N = int(input())
prices = [tuple(map(int, input().split())) for _ in range(N)]

INF = 10**9
memo = [[-1]*3 for _ in range(N)]

def dfs(house, prev):
    if house == N:
        return 0

    if prev != -1 and memo[house][prev] != -1:
        return memo[house][prev]

    res = INF
    for c in range(3):
        if c != prev:
            res = min(res, prices[house][c] + dfs(house+1, c))

    if prev != -1:
        memo[house][prev] = res
    return res

print(dfs(0, -1))
```

```python
# 7	1932	 정수 삼각형
import sys

from io import StringIO
sys.stdin = StringIO('''5
7
3 8
8 1 0
2 7 4 4
4 5 2 6 5''')

input = sys.stdin.readline

N = int(input())
max_sums = [[0] * (N + 1) for _ in range(N + 1)]


for row in range(1, N + 1):
    nums = tuple(map(int, input().split()))
    for col in range(len(nums)):
        upleft = max_sums[row-1][col] + nums[col]
        upright = max_sums[row-1][col+1] + nums[col]
        max_sums[row][col+1] = max(upleft, upright )

print(max(max_sums[N]))
```

```python
# 8	2579	 계단 오르기

import sys

from io import StringIO
sys.stdin = StringIO('''6
10
20
15
25
10
20''')

input = sys.stdin.readline
N = int(input())
first = int(input())


if N == 1:
    print(first)
    sys.exit(0)

# [n][0] =  n번째 밟았을 때 1칸 뛰기가 가능한 최대값. (2칸뛰었거나, 1칸 한번만 뛴 경우)
# [n][1] =  n번째 밟았을 때 1칸 뛰기가 불가능 한 최대값 (1번 을 두 번 뛰어 온 값)

second = int(input())
scores = [[0] * 2 for _ in range(N + 1)]

scores[1] = [first, 0]
scores[2] = [second, first + second]

for n in range(3, N + 1):
    score = int(input())
    scores[n][0] = max(scores[n-2]) + score
    scores[n][1] = scores[n-1][0] + score

print(max(scores[N]))
```

```python
# 12	11053	 가장 긴 증가하는 부분 수열

import sys

from io import StringIO
sys.stdin = StringIO('''9
3 10 2 1 20 4 6 7 8''')

from bisect import bisect_left

input = sys.stdin.readline

N = int(input())
A = list(map(int, input().split()))

tails = []

for x in A:
    i = bisect_left(tails, x)
    if i == len(tails):
        tails.append(x)
    else:
        tails[i] = x

print(len(tails))
```

```python
# 13	11054	 가장 긴 바이토닉 부분 수열

import sys

from io import StringIO
sys.stdin = StringIO('''10
1 5 2 1 4 3 4 5 2 1''')

from bisect import bisect_left

input = sys.stdin.readline
N = int(input())
A = list(map(int, input().split()))

left_nums = [0] * N  # 왼쪽에 이어 붙일 수 있는 숫자 개수
right_nums = [0] * N # 오른쪽에 이어 붙일 수 있는 숫자 개수


for i in range(N):
    for j in range(i):
        if A[j] < A[i]:
            left_nums[i] = max(left_nums[j] + 1, left_nums[i])

for i in range(N-1, -1, -1):
    for j in range(N-1, i, -1):
        if A[j] < A[i]:
            right_nums[i] = max(right_nums[j] + 1, right_nums[i])

result = [left_nums[i] + right_nums[i] + 1 for i in range(N)]
print(max(result))
```

```python
# 14	2565	 전깃줄
import sys

from io import StringIO
sys.stdin = StringIO('''8
1 8
3 9
2 2
4 1
6 4
10 10
9 7
7 6''')


input = sys.stdin.readline
N = int(input())
lines = list(tuple(map(int, input().split())) for _ in range(N))
lines.sort()

nums = [b for _, b in lines]

dp = [1] * N

for i in range(N):
    next = nums[i]
    best = 0
    for j in range(i):
        if nums[j] < next:
            temp = dp[j]
            if best < temp: 
                best = temp
    dp[i] = best + 1

print(N - max(dp))
```
