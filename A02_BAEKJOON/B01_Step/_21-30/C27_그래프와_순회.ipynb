```python
# 1	24479	 알고리즘 수업 - 깊이 우선 탐색 1

import sys
from bisect import insort
sys.setrecursionlimit(10**7)

from io import StringIO
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline

N, M, R = map(int, input().split())

visited = [0]*N

nodes: dict[int, list[int]] = {}

for _ in range(M):
    start, end = map(int, input().split())
    temp = nodes.setdefault(start, [])
    insort(temp, end)
    temp = nodes.setdefault(end, [])
    insort(temp, start)


count = 1
def dfs(R):
    global count

    if visited[R-1]: return 
    visited[R-1] = count
    count += 1
    for node in nodes.get(R, []):
        if visited[node-1]: continue
        dfs(node)

dfs(R)

print("\n".join(map(str, visited)))
```

```python
# 1	24479	 알고리즘 수업 - 깊이 우선 탐색 1

import sys
sys.setrecursionlimit(10**7)

from io import StringIO
from time import perf_counter, perf_counter_ns
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline


N, M, R = map(int, input().split())

# 방문 순서 기록 (1번 정점부터 사용)
visited = [0] * (N + 1)

# 인접 리스트 (가장 효율적인 그래프 저장 방식)
graph = [[] for _ in range(N + 1)]

# 간선 입력
for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

# 문제 조건: 정점 번호가 작은 것부터 방문
for i in range(1, N + 1):
    graph[i].sort()

order = 1
def dfs(v):
    global order
    visited[v] = order
    order += 1
    for nxt in graph[v]:
        if visited[nxt] == 0:
            dfs(nxt)

dfs(R)

# 출력
print("\n".join(map(str, visited[1:])))
```

```python
# 2	24480	 알고리즘 수업 - 깊이 우선 탐색 2
import sys
sys.setrecursionlimit(10**7)

from io import StringIO
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline


N, M, R = map(int, input().split())

nodes : dict[int, list[int]] = {}
visited = [0] * (N + 1)

for _ in range(M):
    start, end = map(int, input().split())
    if not start in nodes:
        nodes[start] = []
    nodes[start].append(end)

    if not end in nodes:
        nodes[end] = []
    nodes[end].append(start)

for val in nodes.values():
    val.sort(reverse=True)


count = 0
def dfs(r):
    global count
    if visited[r]: return
    count += 1
    visited[r] = count

    for node in nodes.get(r, []):
        if visited[node]: continue
        dfs(node)

dfs(R)
print("\n".join(map(str, visited[1:])))
```

```python
# 3	24444	 알고리즘 수업 - 너비 우선 탐색 1

from bisect import insort
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')


input = sys.stdin.readline
N, M, R = map(int, input().split())

visited = [0] * (N + 1)
nodes: dict[int, list[int]] = {}
dq = deque()

for _ in range(M):
    s, e = map(int, input().split())
    insort(nodes.setdefault(s, []), e)
    insort(nodes.setdefault(e, []), s)


def bfs(sn):
    dq.append(sn)
    visited[sn] = 1
    count = 1

    while dq:
        cur = dq.popleft()

        for node in nodes.get(cur, []):
            if not visited[node]:
                count += 1
                visited[node] = count
                dq.append(node)


bfs(R)
print("\n".join(map(str, visited[1:])))
```

```python
# 6	1260	 DFS와 BFS
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''5 5 3
5 4
5 2
1 2
3 4
3 1''')

input = sys.stdin.readline

N, M, V = map(int, input().split())

links = [[] for _ in range(N + 1)]

for _ in range(M):
    s, e = map(int, input().split())
    links[s].append(e)
    links[e].append(s)

for link in links:
    link.sort()


def dfs(start):
    visited = [False] * (N+1)
    out = []

    def recur(n):
        if not visited[n]:
            out.append(n)
            visited[n] = True

            for node in links[n]:
                if not visited[node]:
                    recur(node)

    recur(start)
    return out


def bfs(start):
    visited = [False] * (N+1)
    out = []

    dq = deque([start])
    visited[start] = True 
    
    while dq:
        cur = dq.popleft()
        out.append(cur)
        for node in links[cur]:
            if not visited[node]:
                visited[node] = True
                dq.append(node)

    return out

print(*dfs(V))
print(*bfs(V))
```

```python
# 7	2667	 단지번호붙이기
import sys

from io import StringIO
sys.stdin = StringIO('''7
0110100
0110101
1110101
0000111
0100000
0111110
0111000''')

input = sys.stdin.readline

N = int(input())

apts = [list(map(int, input().rstrip())) for _ in range(N)]

dangi_count = 0
out = []

directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def dfs(row, col):
    apts[row][col] = 0
    apt_count = 1

    for i, j in directions:
        if 0 <= row + i < N and 0 <= col + j < N:
            if apts[row + i][col + j]:
                apt_count += dfs(row + i, col + j)

    return apt_count

for i in range(N):
    for j in range(N):
        if apts[i][j]:
            dangi_count += 1
            out.append(dfs(i, j))

out.sort()
print(dangi_count)
print("\n".join(map(str, out)))
```

```python
# 8	1012	 유기농 배추
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''2
10 8 17
0 0
1 0
1 1
4 2
4 3
4 5
2 4
3 4
7 4
8 4
9 4
7 5
8 5
9 5
7 6
8 6
9 6
10 10 1
5 5''')

input = sys.stdin.readline

T = int(input())

directions = [(+1, 0), (0, +1), (-1, 0), (0, -1)]
out = [0] * T


def BFS(row, col, field, width, height):

    dq = deque([(row, col)])
    field[row][col] = False

    while dq:
        r, c = dq.popleft()

        for dr, dc in directions:
            h, w = r + dr, c + dc
            if 0 <= h < height and 0 <= w < width:
                if field[h][w]:
                    dq.append((h, w))
                    field[h][w] = False


for i in range(T):
    M, N, K = map(int, input().split())
    field = [[False] * M for _ in range(N)]
    cabbages = []

    for _ in range(K):
        c, r = map(int, input().split())
        field[r][c] = True
        cabbages.append((r, c))

    for r, c in cabbages:
        if field[r][c]:
            BFS(r, c, field, M, N)
            out[i] += 1

print("\n".join(map(str, out)))
```

```python
# 9	2178	 미로 탐색
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''2 25
1011101110111011101110111
1110111011101110111011101''')

input = sys.stdin.readline

N, M = map(int, input().split())
maze = [list(map(int, input().rstrip())) for _ in range(N)]
directions = [(-1, 0), (+1, 0), (0, -1), (0, +1)]


def BFS(height, width):
    
    final_node = (height-1, width-1)
    start_node = (0, 0)

    dq = deque([start_node])
    maze[0][0] = 2

    while dq:
        r, c = dq.popleft()
        for dr, dc in directions:
            nr, nc = (r + dr, c + dc)
            if (nr, nc) == final_node:
                return maze[r][c]
            if 0 <= nr < height and 0 <= nc < width and maze[nr][nc] == 1:
                dq.append((nr, nc))
                maze[nr][nc] = maze[r][c] + 1
    else:
        return -1

print(BFS(N, M))
```
