```python
# 1	24479	 알고리즘 수업 - 깊이 우선 탐색 1

import sys
from bisect import insort
sys.setrecursionlimit(10**7)

from io import StringIO
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline

N, M, R = map(int, input().split())

visited = [0]*N

nodes: dict[int, list[int]] = {}

for _ in range(M):
    start, end = map(int, input().split())
    temp = nodes.setdefault(start, [])
    insort(temp, end)
    temp = nodes.setdefault(end, [])
    insort(temp, start)


count = 1
def dfs(R):
    global count

    if visited[R-1]: return 
    visited[R-1] = count
    count += 1
    for node in nodes.get(R, []):
        if visited[node-1]: continue
        dfs(node)

dfs(R)

print("\n".join(map(str, visited)))
```

```python
# 1	24479	 알고리즘 수업 - 깊이 우선 탐색 1

import sys
sys.setrecursionlimit(10**7)

from io import StringIO
from time import perf_counter, perf_counter_ns
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline


N, M, R = map(int, input().split())

# 방문 순서 기록 (1번 정점부터 사용)
visited = [0] * (N + 1)

# 인접 리스트 (가장 효율적인 그래프 저장 방식)
graph = [[] for _ in range(N + 1)]

# 간선 입력
for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

# 문제 조건: 정점 번호가 작은 것부터 방문
for i in range(1, N + 1):
    graph[i].sort()

order = 1
def dfs(v):
    global order
    visited[v] = order
    order += 1
    for nxt in graph[v]:
        if visited[nxt] == 0:
            dfs(nxt)

dfs(R)

# 출력
print("\n".join(map(str, visited[1:])))
```
