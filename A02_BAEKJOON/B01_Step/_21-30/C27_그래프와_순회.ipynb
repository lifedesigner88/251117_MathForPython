```python
# 1	24479	 알고리즘 수업 - 깊이 우선 탐색 1

import sys
from bisect import insort
sys.setrecursionlimit(10**7)

from io import StringIO
sys.stdin = StringIO('''5 5 1
1 4
1 2
2 3
2 4
3 4''')

input = sys.stdin.readline

N, M, R = map(int, input().split())

visited = [0]*N

nodes: dict[int, list[int]] = {}

for _ in range(M):
    start, end = map(int, input().split())
    temp = nodes.setdefault(start, [])
    insort(temp, end)
    temp = nodes.setdefault(end, [])
    insort(temp, start)


count = 1
def dfs(R):
    global count

    if visited[R-1]: return 
    visited[R-1] = count
    count += 1
    for node in nodes.get(R, []):
        if visited[node-1]: continue
        dfs(node)

dfs(R)

print("\n".join(map(str, visited)))
```

```python
# 6	1260	 DFS와 BFS
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''5 5 3
5 4
5 2
1 2
3 4
3 1''')

input = sys.stdin.readline

N, M, V = map(int, input().split())

links = [[] for _ in range(N + 1)]

for _ in range(M):
    s, e = map(int, input().split())
    links[s].append(e)
    links[e].append(s)

for link in links:
    link.sort()


def dfs(start):
    visited = [False] * (N+1)
    out = []

    def recur(n):
        if not visited[n]:
            out.append(n)
            visited[n] = True

            for node in links[n]:
                if not visited[node]:
                    recur(node)

    recur(start)
    return out


def bfs(start):
    visited = [False] * (N+1)
    out = []

    dq = deque([start])
    visited[start] = True 
    
    while dq:
        cur = dq.popleft()
        out.append(cur)
        for node in links[cur]:
            if not visited[node]:
                visited[node] = True
                dq.append(node)

    return out

print(*dfs(V))
print(*bfs(V))
```

```python
# 8	1012	 유기농 배추
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''2
10 8 17
0 0
1 0
1 1
4 2
4 3
4 5
2 4
3 4
7 4
8 4
9 4
7 5
8 5
9 5
7 6
8 6
9 6
10 10 1
5 5''')

input = sys.stdin.readline

T = int(input())

directions = [(+1, 0), (0, +1), (-1, 0), (0, -1)]
out = [0] * T


def BFS(row, col, field, width, height):

    dq = deque([(row, col)])
    field[row][col] = False

    while dq:
        r, c = dq.popleft()

        for dr, dc in directions:
            h, w = r + dr, c + dc
            if 0 <= h < height and 0 <= w < width:
                if field[h][w]:
                    dq.append((h, w))
                    field[h][w] = False


for i in range(T):
    M, N, K = map(int, input().split())
    field = [[False] * M for _ in range(N)]
    cabbages = []

    for _ in range(K):
        c, r = map(int, input().split())
        field[r][c] = True
        cabbages.append((r, c))

    for r, c in cabbages:
        if field[r][c]:
            BFS(r, c, field, M, N)
            out[i] += 1

print("\n".join(map(str, out)))
```

```python
# 9	2178	 미로 탐색
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''2 25
1011101110111011101110111
1110111011101110111011101''')

input = sys.stdin.readline

N, M = map(int, input().split())
maze = [list(map(int, input().rstrip())) for _ in range(N)]
directions = [(-1, 0), (+1, 0), (0, -1), (0, +1)]


def BFS(height, width):
    
    final_node = (height-1, width-1)
    start_node = (0, 0)

    dq = deque([start_node])
    maze[0][0] = 2

    while dq:
        r, c = dq.popleft()
        for dr, dc in directions:
            nr, nc = (r + dr, c + dc)
            if (nr, nc) == final_node:
                return maze[r][c]
            if 0 <= nr < height and 0 <= nc < width and maze[nr][nc] == 1:
                dq.append((nr, nc))
                maze[nr][nc] = maze[r][c] + 1
    else:
        return -1

print(BFS(N, M))
```

```python
# 11	7562	 나이트의 이동
from typing import Deque
from collections import deque
import sys

from io import StringIO
sys.stdin = StringIO('''3
8
0 0
7 0
100
0 0
30 50
10
1 1
1 1''')

input = sys.stdin.readline

steps = (-2, -1, +1, +2)
directions = [
    (dr, dc)
    for dr in steps
    for dc in steps
    if abs(dr) + abs(dc) == 3
]



def is_vaild_idx(row_idx, col_idx, board_length):
    return 0 <= row_idx < board_length and 0 <= col_idx < board_length

def BFS(board_length, start_node, goal_node):
    board = [[False] * board_length for _ in range(board_length)]

    dq = deque([ start_node ])  # node, depth
    board[start_node[0]][start_node[1]] = True
    depth = 0
    while dq:
        for _ in range(len(dq)):
            node= dq.popleft()
            if node == goal_node: 
                return depth
            
            for dr, dc in directions:
                nr, nc = node[0] + dr, node[1] + dc
                if is_vaild_idx(nr, nc, board_length) and not board[nr][nc]:
                    dq.append((nr, nc))
                    board[nr][nc] = True
        depth += 1

    return -1

N = int(input())
out = []
for _ in range(N):

    board_length = int(input())
    start_node = tuple(map(int, input().split()))
    goal_node = tuple(map(int, input().split()))

    out.append(BFS(board_length, start_node, goal_node))

print("\n".join(map(str, out)))

```

```python
# 11	7562	 나이트의 이동

import sys ## 양방향 BFS 심화
from collections import deque

from io import StringIO
sys.stdin = StringIO('''3
8
0 0
7 0
100
0 0
30 50
10
1 1
1 1''')

input = sys.stdin.readline

steps = (-2, -1, 1, 2)
directions = [
    (dr, dc)
    for dr in steps
    for dc in steps
    if abs(dr) + abs(dc) == 3
]

def is_valid(r, c, n):
    return 0 <= r < n and 0 <= c < n

def bidirectional_bfs(n, start, goal):
    if start == goal:
        return 0

    visited_start = [[-1] * n for _ in range(n)]
    visited_goal = [[-1] * n for _ in range(n)]

    dq_start = deque([start])
    dq_goal = deque([goal])

    visited_start[start[0]][start[1]] = 0
    visited_goal[goal[0]][goal[1]] = 0

    while dq_start and dq_goal:
        if len(dq_start) <= len(dq_goal):
            res = expand(dq_start, visited_start, visited_goal, n)
        else:
            res = expand(dq_goal, visited_goal, visited_start, n)

        if res != -1:
            return res

    return -1

def expand(dq, visited_self, visited_other, n):
    for _ in range(len(dq)):
        r, c = dq.popleft()
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if not is_valid(nr, nc, n):
                continue
            if visited_self[nr][nc] != -1:
                continue

            visited_self[nr][nc] = visited_self[r][c] + 1

            if visited_other[nr][nc] != -1:
                return visited_self[nr][nc] + visited_other[nr][nc]

            dq.append((nr, nc))
    return -1

T = int(input())
for _ in range(T):
    n = int(input())
    start = tuple(map(int, input().split()))
    goal = tuple(map(int, input().split()))
    print(bidirectional_bfs(n, start, goal))
```
