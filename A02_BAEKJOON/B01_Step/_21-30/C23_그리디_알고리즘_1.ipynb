```python
# 1	11047	 동전 0
import sys

from io import StringIO
sys.stdin = StringIO('''10 4790
1
5
10
50
100
500
1000
5000
10000
50000''')

input = sys.stdin.readline

N, K = map(int, input().split())

price = K
coins = [int(input()) for _ in range(N)]
coins.reverse()


counter = 0
for coin in coins:
    if coin <= price:
        counter += price // coin
        price = price % coin

print(counter)
```

```python
# 2	1931	 회의실 배정

import sys

from io import StringIO
sys.stdin = StringIO('''11
1 4
3 5
0 6
5 7
3 8
5 9
6 10
8 11
8 12
2 13
12 14''')

input = sys.stdin.readline

N = int(input())

meetings = [tuple(map(int, input().split())) for _ in range(N)]
meetings.sort(key=lambda x: (x[1], x[0]))

count = 0
end_time = 0

for start, end in meetings:
    if start >= end_time:
        count += 1
        end_time = end

print(count)
```

```python
# 3	11399	 ATM
import sys

from io import StringIO
sys.stdin = StringIO('''5
3 1 4 3 2''')

input = sys.stdin.readline

N = int(input())

P = list(map(int, input().split()))
P.sort()

prefix = [0] * (N + 1)

for i in range(1, len(prefix)):
    prefix[i] = prefix[i - 1] + P[i - 1]

print(sum(prefix))
```

```python
# 4	1541	 잃어버린 괄호
import sys

from io import StringIO
sys.stdin = StringIO('''55-50+40''')

input = sys.stdin.readline

# + 나오면 바로 더함, - 는 더하기를 모두 없애고 빼기 즉 더하기 먼저하고 빼기 나중에. 

from collections import deque

eq = input()

num = []
dq = deque()

for c in eq:
    if c.isdigit():
        num.append(c)
    elif c == '-':
        dq.append(int("".join(num)))
        num.clear()
        dq.append(c)
    else: # + 이면
        dq.append(int("".join(num)))
        num.clear()
        dq.append(c)
dq.append(int("".join(num)))

cal_plus = deque()
while dq:
    temp = dq.popleft()
    
    if temp == '+':
        cal_plus.append(cal_plus.pop() + dq.popleft())
    else:
        cal_plus.append(temp)

cal_minus = deque()
while cal_plus:
    temp = cal_plus.popleft()
    if temp == '-':
        cal_minus.append(cal_minus.pop() - cal_plus.popleft())
    else:
        cal_minus.append(temp)

print(cal_minus.pop())


```

```python
# 4	1541	 잃어버린 괄호
import sys

from io import StringIO
sys.stdin = StringIO('''55-50+40''')

input = sys.stdin.readline

s = input().strip()

total = 0
cur = 0
minus_mode = False  # 첫 '-' 이후면 True

for ch in s + '+':          # 마지막 숫자 처리용으로 가짜 연산자 하나 붙임
    if ch.isdigit():
        cur = cur * 10 + (ord(ch) - ord('0'))
    else:  # '+', '-'
        if minus_mode:
            total -= cur
        else:
            total += cur
        cur = 0
        if ch == '-':
            minus_mode = True

print(total)

```

```python
# 4	1541	 잃어버린 괄호
import sys

from io import StringIO
sys.stdin = StringIO('''55-50+40''')

input = sys.stdin.readline

expr = input().strip()
parts = expr.split('-')

ans = sum(map(int, parts[0].split('+')))
for p in parts[1:]:
    ans -= sum(map(int, p.split('+')))

print(ans)
```

```python
# 5	13305	 주유소
import sys

from io import StringIO
sys.stdin = StringIO('''4
2 3 1
5 2 4 1''')

input = sys.stdin.readline

N = int(input())

kms = tuple(map(int, input().split()))
prices = tuple(map(int, input().split()))

end = len(prices) - 1

price = 0

while 0 < end:
    min_price = min(prices[:end])
    end = prices.index(min_price)
    price += min_price * sum(kms[end:])
    kms = kms[0:end]
    
print(price)
```

```python
import sys

from io import StringIO
sys.stdin = StringIO('''4
2 3 1
5 2 4 1''')


input = sys.stdin.readline

N = int(input())
kms = list(map(int, input().split()))      # 길이 N-1
prices = list(map(int, input().split()))   # 길이 N

min_price = prices[0]
total = 0

for i in range(N - 1):
    if prices[i] < min_price:
        min_price = prices[i]
    total += min_price * kms[i]

print(total)
```
