```python
# 1	12852	 1로 만들기 2
import sys
from collections import deque

from io import StringIO
sys.stdin = StringIO('''10''')

input = sys.stdin.readline

def solve():

    N = int(input())

    parent = dict() # {자식(num) : 부모 (num)}
    parent[N] = -1  # (-1) 최상단 노드 & 방문 했다는 표시. 

    dq = deque([N])
    while dq:
        cur_num = dq.popleft()

        if cur_num == 1:
            break

        if cur_num % 3 == 0 and (cur_num // 3) not in parent:
            dq.append(cur_num // 3)
            parent[cur_num // 3] = cur_num

        if cur_num % 2 == 0 and (cur_num // 2) not in parent:
            dq.append(cur_num // 2)
            parent[cur_num // 2] = cur_num

        if (cur_num - 1) not in parent:
            dq.append(cur_num - 1)
            parent[cur_num - 1] = cur_num

    result = [cur_num]
    
    while True:
        parent_num = parent[cur_num]
        if parent_num == -1:
            break
        result.append(parent_num)
        cur_num = parent_num

    result.reverse()
    
    print(len(result) - 1)
    print(*result)


solve()
```
