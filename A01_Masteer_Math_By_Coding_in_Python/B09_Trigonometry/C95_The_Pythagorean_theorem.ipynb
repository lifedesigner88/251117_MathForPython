```python
from matplotlib.pyplot import *
from numpy import max, min, array, sqrt

a = array([0, 0])
b = array([3, 4])
c = array([3, 0])


def line_length(aaa, bbb):
    return sqrt((aaa[0] - bbb[0])**2 + (aaa[1] - bbb[1])**2).round(2)


def try_angle(aa, bb, cc):
    lst = [aa, bb, cc]
    spotname = ["a", "b", "c"]
    for i in range(3):
        first = lst[i]
        second = lst[(i + 1)%3]
        plot([first[0], second[0]], [first[1], second[1]])
        text(first[0] - .2, first[1] - .2, f"({first[0]}, {first[1]})")
        text((first + second)[0]/2 + .1, (first + second)[1]/2 + .1,
             f"{spotname[i]} = {str(line_length(first, second))}")


try_angle(a, b, c)
max_val = max([a, b, c]) + 1
min_val = min([a, b, c]) - 1
axis('square')
xlim(min_val, max_val)
ylim(min_val, max_val)
show()
```

```python
from matplotlib.pyplot import *
from numpy import sqrt, abs, angle, arctan, linspace, pi, tan, sin, arctan2, cos

z = 3 + 4j

mag1 = sqrt(z.real**2 + z.imag**2)
mag2 = abs(z)

print(mag1)
print(mag2)

ang1 = angle(z)
print(ang1, "angle")
ang2 = arctan2(z.imag, z.real)
print(ang2, "arctan2")


theta = linspace(-pi, pi, 100)
plot(theta, tan(theta), color="red", alpha=0.4)
plot(theta, sin(theta))
plot(theta, cos(theta))
axvline(x=0, c='k', ls='--', lw=.5)
axhline(y=0, c='k', ls='--', lw=.5)
axis('square')
xlim(-pi, pi)
ylim(-pi, pi)
show()
```
