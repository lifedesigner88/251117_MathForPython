```python
from matplotlib.pyplot import *
from numpy import random, mean, std

desired_mean = 15
desired_std = 4.3

nums = random.randn(10000)
nums = nums - mean(nums)
nums = nums / std(nums)

nums =  nums * desired_std + 15

print(mean(nums))
print(std(nums))

hist(nums)
axvline(x=15, color='red')
axvline(x=15-desired_std, color='red')
axvline(x=15+desired_std, color='red')
xlim(-50, 50)

# mean(평균) std(표준편차)
```
