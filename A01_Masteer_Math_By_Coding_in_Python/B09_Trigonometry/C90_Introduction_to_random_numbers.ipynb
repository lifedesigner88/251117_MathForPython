```python
from matplotlib.pyplot import *
from numpy import random

nums = random.rand(1000)

minv = 12
maxv = 21

nums = nums*(maxv - minv) + minv

plot(nums, 'o')
show()
hist(nums)
show()
```

```python
from matplotlib.pyplot import *
from numpy import random

nums = random.randn(5000)

minv = 12
maxv = 21

nums = nums*(maxv - minv) + minv

plot(nums, 'o',alpha=.2)
show()
hist(nums)
show()
```
