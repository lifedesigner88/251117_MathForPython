```python
from sympy import symbols, cos, sin, diff, latex
from IPython.display import Math

theta = symbols('theta')

f = sin(theta)

for i in range(4):
    display(Math(
        rf"""
        \frac{{d}}{{dx}} \Big( {latex(f).replace(r"{\left(", " ").replace(r"\right)}", " ")} \Big)
        = {latex(diff(f)).replace(r"{\left(", " ").replace(r"\right)}", "")}
        """))
    f = diff(f)
    print()
```

```python
from sympy.plotting import plot

lst = [f]

n = 3
for _ in range(n):
    lst.append(diff(lst[-1]))

p = plot(*lst, (theta, -3, 3), show=False)

for i in range(n + 1):
    p[i].label = rf"$ {i + 1}. \: {latex(lst[i])} $"

p.legend = True
p.show()
```

```python
from sympy import cos, sin, symbols, diff
from sympy.plotting import plot

a, x = symbols('a, x')

f_ax = cos(x + sin(x)) + a

functions = [f_ax.subs({a: i}) for i in range(0, 4)]
diff_func = [diff(f) for f in functions]

plot(*functions)
plot(*diff_func)
```
