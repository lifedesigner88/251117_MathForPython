```python
from IPython.display import Math
from sympy import symbols, diff, Function, Eq, latex

x, g = symbols('x g')
fx = x**2
dfx = diff(fx)
f = Function('f')(x)

display(Math(rf' {latex(Eq(f, fx))} , \qquad  {latex(Eq(diff(f, x), diff(fx, x)))}  '))
print()

# Leibniz notation 라이프니츠 로테이션
display(Math(rf' {latex(Eq(f, fx))} , \qquad  \frac{{df}}{{dx}} =  {latex(diff(fx, x))}  '))
print()

# Lagrange notation 라그랑주 로테이션
display(Math(rf" {latex(Eq(f, fx))} , \qquad  f' =    {latex(diff(fx, x))}  "))
print()

# Nweton notation 뉴턴 로테이션
display(Math(rf' {latex(Eq(f, fx))} , \qquad  \dot f  =    {latex(diff(fx, x))}  '))
```

```python
from sympy.plotting import plot
from sympy import symbols, diff, latex, Eq, Function

x = symbols('x')
f = Function('f')(x)
fx = 3 - x**3

p = plot(fx, diff(fx), (x, -3, 3), show=False, size=(7, 8))
p[0].label = rf'$ {latex(Eq(f, fx))} $'
p[1].label = rf'$ {latex(Eq(f, diff(fx)))} $'
p.ylim = [-10, 10]
p.legend = True
p.show()
```
