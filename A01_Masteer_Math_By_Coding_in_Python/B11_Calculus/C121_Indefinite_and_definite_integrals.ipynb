```python
from sympy import symbols, integrate, Function, Eq
from sympy.plotting import plot

x = symbols('x')
fx = Function('f')(x)

f = x
i_f = integrate(f, x)
display(Eq(fx, f))
display(Eq(integrate(fx), i_f))
plot(f, i_f, (x, -5, 5), show=False, size=(5, 7)).show()
```

```python
f1 = x**3/(x - 2)

p = plot(f1, integrate(f1), (x, -0.1, 4), show=False)
p.ylim = (-100, 100)
p.show()
```

```python
from sympy import sin, Function, Eq, diff, integrate, latex
from IPython.display import Math

e_fx = 2*x**3 + sin(x)

f = Function('f')(x)

display(Eq(f, e_fx))
display(Math(rf"""
    f(x)' = {latex(diff(e_fx))}
    """))
display(Math(rf"""
    \int f(x)' dx = \int {latex(diff(e_fx))} =  {latex(e_fx)} + C
    """))


```
