```python
from sympy import diff, symbols, Function, Eq

x, y = symbols('x, y')

f = x**2 + x*y**2

fxy = Function('f')(x, y)

display(Eq(fxy, f))

print()
display(Eq(diff(fxy, x), diff(f, x)))

print()
display(Eq(diff(fxy, y), diff(f, y)))
```

```python
# %matplotlib notebook

from sympy.plotting import plot3d

plot3d(f, (x, -3, 3), (y, -3, 3))
plot3d(diff(f, x), (x, -3, 3), (y, -3, 3))
plot3d(diff(f, y), (x, -3, 3), (y, -3, 3))
```
