```python
from sympy import symbols, diff, lambdify, exp, latex
from numpy import linspace, arange
from IPython.display import Math

x, a = symbols('x, a')
a_s = arange(0, 2.25, .25)
f_xa = x**2*exp(-a*x**2)
df_xa = diff(f_xa, x)


lamb_fxa = lambdify((x, a), f_xa)
lamb_df_xa = lambdify((x, a), df_xa)

for i in a_s:
    f_x = f_xa.subs(a, float(i))
    df_x = df_xa.subs(a, i)
    if df_x.subs(x, 1) == 0 :
        display(Math(rf" \Rightarrow {latex(f_x)} \quad \text{{has a critical point at}} \quad x = 1"))
    elif df_x.subs(x, 2) == 0 :
        display(Math(rf" \Rightarrow  {latex(f_x)} \quad \text{{has a critical point at}} \quad x = 2"))
    else:
        display(Math(rf" \quad \; {latex(f_x)}"))
```

```python
from matplotlib.pyplot import *

xs = linspace(-3, 3, 100)

figure(figsize=(15, 8))

subplot(121)
for i in a_s:
    plot(xs, lamb_fxa(xs, i))
axvline(x=1, c='k', ls='--', lw=.5)
axvline(x=2, c='k', ls='--', lw=.5)
xlim(-3, 3)
ylim(0, 2)

subplot(122)
for i in a_s:
    plot(xs, lamb_df_xa(xs, i))
xlim(-3, 3)
ylim(-1.5, 1.5)
axhline(y=0, c='k', ls='--', lw=.5)
axvline(x=1, c='k', ls='--', lw=.5)
axvline(x=2, c='k', ls='--', lw=.5)
show()
```
