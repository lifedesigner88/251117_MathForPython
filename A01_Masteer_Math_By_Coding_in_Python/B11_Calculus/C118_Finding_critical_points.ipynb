```python
from matplotlib.pyplot import *
from numpy import linspace, exp, diff
from scipy.signal import find_peaks

x = linspace(-5, 5, 1001)
fx = x**2*exp(-x**2)
dfx = diff(fx)/diff(x)

plot(x, fx)
plot(x[0:-1], dfx)


def peaks_dots(fun):
    max_peaks = find_peaks(fun)[0]
    min_peaks = find_peaks(-fun)[0]
    peak_indexs = [*max_peaks, *min_peaks]

    peak_xs = x[peak_indexs]
    peak_ys = fun[peak_indexs]

    plot(peak_xs, peak_ys, 'ro')


peaks_dots(fx)
peaks_dots(dfx)
show()
```
