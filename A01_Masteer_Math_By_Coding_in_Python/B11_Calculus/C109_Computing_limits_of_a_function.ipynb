```python
from matplotlib.pyplot import *
from IPython.display import Math
from sympy import symbols, limit, Limit, latex, Eq, lambdify
from numpy import linspace

x = symbols('x')

fx = x**3

lim_pnt = 1.5
lim = limit(fx, x, lim_pnt)
# display(Math(r' \lim_{x \to %g} %s = %g '%(lim_pnt, latex(fx), lim)))
# display(Math(rf' \lim_{{ x \to {lim_pnt} }} {latex(fx)} = {lim.round(3)} '))

limexp = Limit(fx, x, lim_pnt)
eq = Eq(limexp, limexp.doit())
display(Math(latex(eq).replace('^+', "")))

lamb_fx = lambdify(x, fx)
input_x = linspace(-5, 5, 200)

plot(input_x, lamb_fx(input_x))
plot(lim_pnt, lim, 'ro')
show()

```

```python
fx2 = x**2 / (x-2)

lim_fx_exp = Limit(fx2, x, 2, dir="+")
lim_fx_val = lim_fx_exp.doit()
display(Eq(lim_fx_exp, lim_fx_val))

lim_fx_exp = Limit(fx2, x, 2, dir="-")
lim_fx_val = lim_fx_exp.doit()
display(Eq(lim_fx_exp, lim_fx_val))

lamb_fx2 = lambdify(x, fx2)
input_x = linspace(-1, 3, 200)
plot(input_x, lamb_fx2(input_x))
ylim(-200,200)
show()
```
