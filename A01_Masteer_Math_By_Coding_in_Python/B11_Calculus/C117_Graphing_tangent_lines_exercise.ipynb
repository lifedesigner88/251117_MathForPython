```python
from sympy import symbols, Expr, diff
from sympy.plotting import plot
from numpy import linspace
x = symbols('x')
f = x**2


def draw_tanline(fx: Expr, a: float):
    f_a = fx.subs(x, a)
    df_a = diff(fx).subs(x, a)
    return df_a*(x - a) + f_a

lines = [draw_tanline(f, i) for i in linspace(-2,2,200)]

p = plot(*lines,(x, -2, 2), show=False, axis=False, line_color='k', size=(6,7))
p.ylim = (-1, 4)
p.show()
```
