```python
from sympy import symbols, diff

# f(x)에서  x = a  에서의 접선의 방정식 g(x) = f'(a)*(x-a) + f(a)
x = symbols('x')


def tanline(fx, a):
    df_a = diff(fx).subs(x, a)
    f_a = fx.subs(x, a)
    return df_a*(x - a) + f_a


f_x = x**2
g_x = tanline(f_x, 1)

from sympy.plotting import plot

p = plot(f_x, g_x, (x, -2, 2), show=False, size=(6, 7))
p.ylim=(-2, 4)
p.show()
```

```python
from matplotlib.pyplot import *
from sympy import lambdify, latex
from numpy import linspace


xs = linspace(-2, 2, 100)
lamb_fx = lambdify(x, f_x)(xs)
lamb_gx = lambdify(x, g_x)(xs)

figure(figsize=(6,7))
plot(xs, lamb_fx, label='$f(x) = x^2$' )
plot(xs, lamb_gx, label=f"$g(x) = {latex(g_x)} $")
plot(1, f_x.subs(x, 1), 'ro')
axhline(y=0, c='k', lw=.5, ls='--')
axvline(x=0, c='k', lw=.5, ls='--')
ylim(-2, 4)
legend()
show()
```
