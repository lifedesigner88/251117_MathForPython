```python
from numpy import zeros, arange

def create_toeplitz(lst1, lst2=None):
    length = len(lst1)

    if lst2 is None:
        z = zeros((length, length), dtype=int)
        for i in range(length):
            for j in range(i, length):
                z[i, j] = lst1[j - i]
        return z + z.T - z*eye(length, dtype=int)

    elif len(lst2) != length or lst1[0] != lst2[0]:
        raise ValueError("list langth must be equal, or list's first value must be same")

    else:
        z1 = zeros((length, length), dtype=int)
        z2 = zeros((length, length), dtype=int)

        for i in range(length):
            for j in range(i, length):
                z1[i, j] = lst1[j - i]
                z2[i, j] = lst2[j - i]

        return z1 + z2.T - z1*eye(length, dtype=int)

toe1 = create_toeplitz([1, 4, 2, 3], [1, 7, 4, 5])
toe2 = create_toeplitz(arange(1,10))

imshow(toe2)
show()
```

```python
from scipy.linalg import toeplitz
toe3 = toeplitz(arange(1,10))
print()
print(toe3)
```
