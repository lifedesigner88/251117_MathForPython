```python
from matplotlib.pyplot import *
from numpy import array

o = array([0, 0])
a = array([1, 2])

a2 = 2*a

plot(*zip(o, a), 'b-', alpha=.5)
plot(*zip(o, a2), 'r--', alpha=.5)

c = complex(*list(a))
length_a = abs(c)
print(length_a)

c = complex(*list(a2))
length_a2 = abs(c)
print(length_a2)

print(length_a2/length_a)

length = 5
axis('square')
grid()
ylim(-length, length)
xlim(-length, length)
show()
```

```python
from matplotlib.pyplot import *
from numpy import diag
from numpy.random import randint
from numpy.linalg import eig

M = randint(-5, 6, (5, 5))
D = M@M.T


eigvals, eigvecs = eig(D)

dic = {
    "metrix": D,
    "eigenvalues": diag(eigvals),
    "eigenvectors": eigvecs
}

fig, ax = subplots(1, 3, figsize=(8, 4))

for idx, (key, val) in enumerate(dic.items()):
    ax[idx].set_title(key)
    ax[idx].imshow(val)
    ax[idx].axis('off')

show()
```

```python
target = 1

Mv = D@eigvecs[:, target]
Iv = eigvals[target]*eigvecs[:, target]

print(eigvecs[:, target])
print(eigvals[target])
print(D)
print(Mv)
print(Iv)
```
