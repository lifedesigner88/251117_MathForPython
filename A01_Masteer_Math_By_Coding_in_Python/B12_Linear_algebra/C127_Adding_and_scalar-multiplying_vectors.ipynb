```python
from matplotlib.pyplot import *
from numpy import array, zeros

zero = zeros(2)
v1 = array([-1, 2])
v2 = array([1, 1])

plot(*zip(zero, v1), label="v1")
plot(*zip(zero + v1, v2 + v1), label="v2")
plot(*zip(zero, v1 + v2), label="v1 + v2")

axis("square")
xlim(-2, 2)
ylim(-2, 4)
grid()
legend()
show()
```
