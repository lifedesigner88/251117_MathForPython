```python
from numpy import random, linalg
from matplotlib.pyplot import *

m = 5
A = random.randint(-5, 6, (m, m))
A_inv = linalg.inv(A)

fig, ax = subplots(1, 3, figsize=(9, 5))

ax[0].imshow(A)
ax[0].set_title('A')

ax[1].imshow(A_inv)
ax[1].set_title('A$^{-1}$')

ax[2].imshow(A@A_inv)
ax[2].set_title('A@A$^{-1}$')

for i in ax: i.axis('off')
show()
```

```python
from numpy.linalg import pinv

m = 5
A = random.randint(-m, m + 1, (m, m + 2))

A[:, 0] = A[:, 2]
A[:, 1] = A[:, 2]

A_pinv = pinv(A)
fig, ax = subplots(1, 3, figsize=(9, 5))

dic = {
    "A": A,
    "A*": A_pinv,
    "AA*": A@A_pinv
}

for idx, (key, val) in enumerate(dic.items()):
    ax[idx].imshow(val)
    ax[idx].set_title(key)
    ax[idx].axis('off')

show()
```
