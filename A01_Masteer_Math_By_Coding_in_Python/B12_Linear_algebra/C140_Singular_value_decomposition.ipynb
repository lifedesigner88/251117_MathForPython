```python
from matplotlib.pyplot import *
from numpy.linalg import svd
from numpy import array, diag
from PIL import Image

A = array(Image.open("image.jpg").convert("L"))
print(A.shape)
U, s, V = svd(A, full_matrices=False)

fig, ax = subplots(2, 2, figsize=(9, 11))
ax = ax.ravel()

ax[3].imshow(A, cmap="gray")
ax[3].set_title("Original")
ax[3].axis('off')

k = [10, 30, 70]
for i in range(len(k)):
    
    a = U[:, :k[i]].shape
    b = diag(s[:k[i]]).shape
    c = V[:k[i], :].shape

    print(a, b,  c )
    print((a[0]*a[1] + b[0]*b[1] + c[0]*c[1])/(A.shape[0] * A.shape[1]) * 100) 

    A_k = U[:, :k[i]] @ diag(s[:k[i]]) @ V[:k[i], :]
    ax[i].imshow(A_k, cmap="gray")
    ax[i].set_title(f"Rank {k[i]} Approxi")
    ax[i].axis('off')

show()
```
