```python
from matplotlib.pyplot import *
from numpy import array

M = array([
    [.5, +1],
    [+1, .5]
])

o = array([0, 0])
v = array([1, 2])

Mv = M@v

plot(*zip(o, v), label='v')
plot(*zip(o, Mv), 'r--', label='Mv',)
axis('square')
xlim(-5, 5)
ylim(-5, 5)
grid()
legend()
show()
```

```python
from matplotlib.pyplot import *
from numpy import array, linspace

M = array([
    [.5, +1],
    [+1, .5]
])

x = linspace(-2, 2, 40)

o = array([0, 0])


figure(figsize=(8,8))

for a in x:
    v = array([1, a])
    abs_a = abs(a)
    plot(*zip(o, v) , c=[abs_a/2, .7, abs_a/2], alpha=0.8)
    plot(*zip(o, M@v), c=[abs_a/2, .2, .4])
    

axis('square')
axis('off')
xlim(-3, 3)
ylim(-3, 3)
show()
```
