```python
from sympy import Matrix
from numpy import array

row_vec = array([
    [4, 2]
])
col_vec = array([
    [-2],
    [3]
])

display(Matrix(row_vec))
display(Matrix(col_vec))
```

```python
from matplotlib.pyplot import *

plot([0, row_vec[0, 0]], [0, row_vec[0, 1]])
plot([0, col_vec[0, 0]], [0, col_vec[1, 0]])
axis("square")
grid()
xlim(-5, 5)
ylim(-5, 5)
show()
```

```python
from matplotlib.pyplot import *
from numpy import array

origin = array([0, 0, 0])
vec1 = array([3, 1, 4])
vec2 = array([-1, -1, 2])

vector1 = list(zip(origin, vec1))
vector2 = list(zip(origin, vec2))

fig = figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.plot(*vector1)
ax.plot(*vector2)

ax.quiver(*origin, *vec1, color='b', lw=2, label='v1')
ax.quiver(*origin, *vec2, color='r', lw=2, label='v2')
ax.quiver(*vec2, *(vec1-vec2), color='g', lw=2, label='v1-v2')

ax.plot([-5, 5], [0, 0], [0, 0], ls='--', color='r', alpha=0.5)
ax.plot([0, 0], [-5, 5], [0, 0], ls='--', color='g', alpha=0.5)
ax.plot([0, 0], [0, 0], [-5, 5], ls='--', color='skyblue', alpha=0.5)
ax.plot(*origin, 'k.', )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
show()
```
