```python
from numpy import random, shape, transpose

r = random.randint(1, 10, (3,5))

print(shape(r))
print(shape(r.T))
print(r.T)
print(transpose(r))
```

```python
from matplotlib.pyplot import *
from numpy import ndenumerate

fig, axis = subplots(1, 2)

axis[0].imshow(r)
axis[0].set_title('M')
axis[1].imshow(r.T)
axis[1].set_title('M$^T$')

for (i, j), val in ndenumerate(r):
    axis[0].text(j, i, str(val), ha='center', va='center')

for (i, j), val in ndenumerate(r.T):
    axis[1].text(j, i, str(val), ha='center', va='center')

for ax in axis:
    ax.axis('off')

show()
```
