```python
from matplotlib.pyplot import *
from sympy.plotting import plot, plot_implicit
from sympy import symbols, Eq, And

x, y = symbols("x, y")

eq1 = 3*x - y
eq2 = -x + y
eq3 = 1.5*x - 0.5*y


p1 = plot_implicit(Eq(eq1, 6), (x, 0, 10), (y, 0, 10),
                   line_color='red', show=False)
p1.extend(plot_implicit(Eq(eq2, 2), (x, 0, 10), (y, 0, 10), show=False))
p1.show()

p1 = plot_implicit(Eq(eq1, 6), (x, 0, 10), (y, 0, 10),
                   line_color='red', show=False)
p1.extend(plot_implicit(Eq(eq3, 3), (x, 0, 10), (y, 0, 10), show=False))
p1.show()

p1 = plot_implicit(Eq(eq1, 6), (x, 0, 10), (y, 0, 10),
                   line_color='red', show=False)
p1.extend(plot_implicit(Eq(eq3, 2), (x, 0, 10), (y, 0, 10), show=False))
p1.show()
```
