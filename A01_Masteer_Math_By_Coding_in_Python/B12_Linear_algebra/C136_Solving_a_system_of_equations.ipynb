```python
from numpy import random
from numpy.linalg import inv
from sympy import Matrix, latex
from sympy.abc import a, b, c
from IPython.display import Math

X = random.randint(1, 10, (3, 3))
y = random.randint(1, 10, (3, 1))

beta = Matrix([a, b ,c])

def dm(m): display(Math(m))
def lm(m): return latex(Matrix(m))

dm(rf"{lm(X)} \times {lm(beta)} = {lm(y)} ")
dm(rf" \qquad \qquad \qquad {lm(beta)} = {lm(X)}^{{-1}} \times \:\; {lm(y)} ")
dm(rf" \qquad \qquad \qquad {lm(beta)} = {lm(inv(X).round(2))} \:\; \times \:\; {lm(y)} ")
dm(rf" \qquad \qquad \qquad {lm(beta)} = {lm((inv(X)@y).round(2))} ")
dm(rf"{lm(X)} \times {lm((inv(X)@y).round(2))} = {lm(y)} ")
dm(rf"{lm( (X@(inv(X)@y)).astype(int))}  = {lm(y)} ")
```
