```python
from matplotlib.pyplot import *
from numpy import random, outer

v1 = random.randn(50)
v2 = random.randn(80)

op1 = outer(v1, v2)
op2 = outer(v2, v1)

fig, axis = subplots(1, 2, figsize=(10, 5))
axis[0].imshow(op1)
axis[1].imshow(op2)
show()
```

```python
from numpy import arange, outer

v = arange(1, 11)
w = arange(1, 6)

print(outer(v, w))
print(outer(w, v))


```
