```python
from numpy import random, shape, dot, zeros
from sympy import Matrix, latex
from IPython.display import Math

A = random.randint(1, 10, (3, 4))
B = random.randint(1, 10, (4, 5))


def pm(m): return latex(Matrix(m))


def dm(m): return display(Math(m))


print()

dm(rf"{pm(A)} \times {pm(B)} = {pm(A@B)}")


def mul_two_matrix(head, foot):
    h_shape = shape(head)
    f_shape = shape(foot)

    if h_shape[1] != f_shape[0]:
        raise ValueError('Matrix must be square')

    row = h_shape[0]
    col = f_shape[1]

    result = zeros((row, col), dtype=int)

    for i in range(row):
        for j in range(col):
            result[i, j] = dot(head[i, :], foot[:, j])

    return result


print()

dm(rf"{pm(mul_two_matrix(A, B) - A@B)}")
```
