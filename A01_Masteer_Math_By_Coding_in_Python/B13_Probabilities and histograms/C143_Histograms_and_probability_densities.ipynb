```python
from matplotlib.pyplot import *
from numpy import arange, histogram

data = [1, 2, 3, 4, 5, 2, 6, 1, 2, 3, 5, 2, 1, 3, 4, 9, 8, 7, 6, 6]
# a = hist(data, bins=arange(1, 11))
# show()

y, x = histogram(data, bins=arange(1, 11))
yP = y/sum(y)
bar(x[:-1], yP)
xticks(range(1, 10))
show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, diff

x = linspace(-3, +3, 10000)

fig, ax = subplots(2, 3, figsize=(10, 6))

for i in range(3):

    y = x**(i+1)
    h = histogram(y, 100)
    yP = h[0]/sum(h[0])

    ax[0, i].plot(x, y)
    ax[0, i].set_title(f'$y=x^{i+1}$')

    ax[1, i].plot(h[1][:-1], yP)
show()
```
