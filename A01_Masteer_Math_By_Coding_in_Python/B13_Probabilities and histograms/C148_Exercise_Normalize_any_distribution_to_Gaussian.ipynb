```python
from matplotlib.pyplot import *
from numpy import arctanh
from scipy.stats import rankdata
from numpy.random import randn

n = 500

orig_data = 2**randn(n)

rank_data = rankdata(orig_data)

# norm to [-1 , 1]
norm_data = 2*(rank_data)/(max(rank_data) + 1) - 1

trans_data = arctanh(norm_data)

fig, ax = subplots(1, 3, figsize=(9, 4))

for i, val in enumerate([orig_data, norm_data, trans_data]):
    ax[i].hist(val, 30)
    ax[i].set_xlabel("Value")
    ax[i].set_ylabel("Count")

show()
```
