```python
from matplotlib.pyplot import *
from numpy.random import randn
from seaborn import jointplot, set_theme, kdeplot
n = 1000
x = randn(n)
y = randn(n)**2

hist2d(x, y, 30, vmax=10)
colorbar()
xlabel('x')
ylabel('$x^2$')
show()

set_theme()
jointplot(x=x , y=y, kind='hist', color=[.8, .3, .9]).plot_joint(kdeplot)
xlabel('x')
ylabel('$x^2$')
show()
```
