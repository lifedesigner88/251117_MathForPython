```python
from matplotlib.pyplot import *
from numpy import linspace, pi, sin
from numpy.random import rand

n = 4000

sig1 = sin(linspace(0, 6*pi, n))
sig2 = rand(n)*2 - 1
combi = sig1 + sig2

dic = {
    "Sine wave": sig1,
    "Unifrom numbers": sig2,
    "Combined signal": combi
}

fig, ax = subplots(2, 3, figsize=(9, 5))
ax = ax.ravel()

for idx, (key, val) in enumerate(dic.items()):
    ax[idx].plot(val, '.', alpha=.2)
    ax[idx].set_title(key)
    ax[idx+3].hist(val, 100)

for i in ax:
    i.axis('off')

show()
```

```python
from seaborn import set_theme, histplot

set_theme()
histplot(combi, bins=100, kde=True)
show()
```

```python
from matplotlib.pyplot import *
from numpy import arange, zeros, linspace, sin, histogram
from numpy.random import rand

ns = arange(50, 10000, 100)
nbins = 50

allhist = zeros((len(ns), nbins))

for i in range(len(ns)):
    sig1 = sin(linspace(0, 6*pi, ns[i]))
    sig2 = rand(ns[i])*2 - 1

    allhist[i, :], x = histogram(sig1 + sig2, nbins, density=True)

pcolor(x[:-1], ns, allhist, vmax=.6, cmap='Purples')
xlabel('Data Value')
ylabel('Points in the sample')
colorbar()
show()
```
