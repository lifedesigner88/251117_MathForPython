```python
from matplotlib.pyplot import *
from numpy.random import randn, rand

shift = 5
stretch = 2
n = 10000

pnts = randn(n)*stretch + shift
pnts2 = rand(n)*stretch + shift - stretch/2

fig, ax = subplots(2, 2, figsize=(9, 9))
ax = ax.ravel()
ax[0].plot(pnts, '.', alpha=.4)
ax[0].axhline(5, c='r', ls='--', alpha=.5)

ax[1].hist(pnts, 100)
ax[1].axvline(5, c='r', ls='--', alpha=.5)

ax[2].plot(pnts2, '.', alpha=.4)
ax[2].axhline(5, c='r', ls='--', alpha=.5)

ax[3].hist(pnts2, 100),
ax[3].axvline(5, c='r', ls='--', alpha=.5)

show()
```

```python
from matplotlib.pyplot import *
from numpy.random import poisson
from numpy import mean, var

lam = 5
n = 10000

pnts2 = poisson(lam, n)
print(mean(pnts2), var(pnts2))
fig, ax = subplots(1, 2, figsize=(9, 5))

ax[0].plot(pnts2, 's', alpha=.2)
ax[0].axhline(5, c='r', ls='--',)

ax[1].hist(pnts2, bins=range(0, max(pnts2) + 1), ec='w')
ax[1].axvline(5, c='r', ls='--', alpha=.5)

show()
```
