```python
from matplotlib.pyplot import *
from numpy.random import rand
from numpy import zeros

faircoin = .5
biascoin = .6

nTrials = 10

results = zeros((2, nTrials))

for i in range(nTrials):
    for j, val in enumerate([faircoin, biascoin]):
        results[j, i] = rand() > val

plot(results[0] - 0.05, 's', markersize=10)
plot(results[1] + 0.05, 's', markersize=10)
yticks([0, 1], ['Tails', 'Heads'])
ylim(-0.5, 1.5)
show()
```

```python
from pandas import DataFrame
df = DataFrame(results, index=['fair', 'bias'], dtype='int')
df['Ave.'] = df.mean(axis=1)
display(df)
```

```python
from matplotlib.pyplot import *
from numpy import array, histogram, zeros
from numpy.random import randint

theo_data = array([1, 1, 2, 3, 4, 5, 6])


y, x = histogram(theo_data, bins=range(1, 8))

tx = x[:-1]
ty = y/7

plot(tx, ty, 'ms-', markersize=8, label="Theoretical")

ntry = 1000
results = zeros(ntry)

for i in range(ntry):
    num = randint(7)
    if num == 0:
        num = 1
    results[i] = num

y, x = histogram(results, bins=range(1, 8))

ex = x[:-1]
ey = y/ntry

bars = bar(ex, ey, label="Empirical")
bar_label(bars, labels=[f"{v*100:.0f}%" for v in ey],
          padding=-15, color='white')

legend()
ylim(0, 0.37)
show()
```
