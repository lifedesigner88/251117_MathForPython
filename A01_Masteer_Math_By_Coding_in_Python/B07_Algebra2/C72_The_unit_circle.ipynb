```python
from numpy import linspace, sin, cos, pi
from matplotlib.pyplot import *

x = linspace(0, 2*pi, 100)

plot(cos(x), sin(x), 'b')
plot([0, cos(pi/4)], [0, sin(pi/4)], 'r--', lw=0.7)
axvline(0, c=[.8, .8, .8], ls='--', lw=1)
axhline(0, c=[.8, .8, .8], ls='--', lw=1)
axis('square')
show()

plot(x, sin(x), 'r')
plot(x, cos(x), 'g')
axis('square')
xlim(0, 2*pi)
ylim(-2, +2)
axhline(0, c='k', ls='--', lw=1)
show()
```

```python
from matplotlib.pyplot import *
from random import randint

from numpy import linspace, sin, cos, e, exp

x = linspace(-3, 1.83, 50)
f = exp(x)

for i in f:
    plot([0, cos(i)], [0, sin(i)], c=[(randint(0, 9)/10) for _ in range(3)])

axis('square')
axis('off')
show()
```
