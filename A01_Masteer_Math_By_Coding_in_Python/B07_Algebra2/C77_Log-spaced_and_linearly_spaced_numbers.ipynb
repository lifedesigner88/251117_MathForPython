```python
from matplotlib.pyplot import *
from numpy import log, linspace, logspace, log10

a = 2
b = 100000
n = 50

x = linspace(a, b, n)
l = logspace(log10(a), log10(b), n)

plot(x, x, label='linear')
plot(x, l, label='log')

axis('square')
axvline(x=0, c='k', ls='--', lw=0.5)
axhline(y=0, c='k', ls='--', lw=0.5)
legend()
show()
```
