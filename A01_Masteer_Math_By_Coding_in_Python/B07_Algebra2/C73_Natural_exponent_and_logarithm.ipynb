```python
from numpy import linspace, exp, log, e
from matplotlib.pyplot import *

x = linspace(0, 3, 100)
y = exp(x)
xlabel('x')
ylabel('$e^x$', fontsize=14)
plot(x, y)
show()

x = linspace(1, e, 100)
y = log(x)
xlabel('x')
ylabel('$ log(x) $', fontsize=14)
plot(x, y)
show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, exp, log

x = linspace(0.0001, 10, 10)
y1 = log(exp(x))
y2 = exp(log(x))

plot(x, y1, 'b', label=r'$\log(e^x)$')
plot(x, y2, 'ro', label=r'$e^{\log(x)}$')
legend()
show()
```
