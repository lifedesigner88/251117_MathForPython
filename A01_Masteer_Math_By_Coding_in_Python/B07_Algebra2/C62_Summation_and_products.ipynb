```python
import numpy as np
```

```python
from math import prod

lst = [1, 3, 4, 1, 6]

print(sum(lst))
print(prod(lst))

print(np.sum(lst))  # 모두 더하기

print(np.prod(lst))  # 모두 곱하기

print(np.cumsum(lst))  # 누적 더하기
print(np.cumprod(lst))  # 누적 곱하기

```

```python
from matplotlib import pyplot as plt

plt.plot(lst, 'rs-')
plt.plot(np.cumsum(lst), 'bo-')

plt.legend(['Summation', 'Cumulative Sum'])
plt.show()
```

```python
a = np.arange(1,6)

summation = np.sum(a)
squaresum = np.sum([v**2 for v in a])
print(summation/squaresum == 1/summation)

product = np.prod(a)
cumulateProduct = np.prod([v**2 for v in a])

print(product / cumulateProduct == 1/product)
```
