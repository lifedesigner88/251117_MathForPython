```python
from matplotlib.pyplot import *
from numpy import linspace, pi, cos, sqrt, max, argmax

x = linspace(0, 12*pi, 2000)
fx = -cos(x) - sqrt(x)


def find_peaks(lis2):
    peaks = []
    for i in range(len(lis2)):
        if i == 0 or i == len(lis2) - 1:
            continue
        else:
            if lis2[i - 1] < lis2[i] and lis2[i] > lis2[i + 1]:
                peaks.append(i)
    return peaks


plot(x, fx)
peaks_x = find_peaks(fx)
peaks_y = -cos(x[peaks_x]) - sqrt(x[peaks_x])
plot(x[peaks_x], peaks_y, 'ro')
show()
```

```python
from numpy import argmin, min

x = linspace(-2, 2, 1001)
fx2 = x**3 + x**4

xlim(-1.2, .5)
ylim(-.2, .3)
plot(x, fx2)
plot(x[argmin(fx2)], min(fx2), 'ro')
show()
```
