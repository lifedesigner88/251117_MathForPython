```python
from numpy import arange

a = 2
d = 3
maxN = 10
n = arange(0, maxN)

ariseq = a + d*n
print(ariseq)

a = 2
r = 3
n = arange(0, maxN)

geoseq = a*r**n
print(geoseq)
```

```python
from matplotlib.pyplot import *

plot(n, ariseq, 'ks', label='Arithmetic')
plot(n, geoseq, 'ro', label='Geometric')
legend()
```
