```python
from matplotlib.pyplot import *
from numpy import log, sqrt, exp, linspace

h = 1
t = linspace(-3, 3, 1001)

# gaussian
g = exp(-4*log(2)*t**2/h**2)

# plot
plot(t, g)
xlabel('t')
ylabel('y')
title('Gaussian with FWHM=' + str(h))

yVal2find = .5
tvals = h*sqrt(log(yVal2find)/(-4*log(2)))
print(tvals, -tvals)
```

```python
from sympy import symbols, exp, log, latex, Eq, solve
from IPython.display import display, Math

t, h = symbols('t h', real=True, positive=True)
g = exp(-4*log(2)*t**2/h**2)

display(Math(r'\Large f(x) = %s'%latex(g).replace('log', 'ln')))

result = solve(Eq(g, 1/2), t)
display(result[0])
```

```python
from matplotlib.pyplot import *
from numpy import log, exp, zeros

h = linspace(0.01, 1, 100)
t = linspace(-3, 3, 1001)

M = zeros((len(h), len(t)))

for i in range(len(h)):
    for j in range(len(t)):
        M[i, j] = exp(-4*log(2)*t[j]**2/h[i]**2)


pcolormesh(M)
show()

imshow(M, origin='lower')
show()
```
