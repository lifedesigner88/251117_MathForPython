```python
from matplotlib import pyplot as plt

z = complex(2, 3)

plt.plot(z.real, z.imag, 'ro')
plt.plot([0, z.real], [0, z.imag], 'r')
plt.axvline(0, color='k', linestyle='--', linewidth=1)
plt.axhline(0, color='k', linestyle='--', linewidth=1)

plt.plot()
plt.xlabel('Real Number')
plt.ylabel('Imaginary Number')
plt.axis('square')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
```

```python
from matplotlib import pyplot as plt

lst = [-3 + 1j, -1 + 1j]
lineColor = ['r', 'g', 'b', 'm']


def drawline(com: complex, color: str, label: str):
    plt.plot([0, com.real], [0, com.imag], color, label=label)


for i in range(2):
    drawline(lst[i], lineColor[i], f"z{i+1}")

drawline(lst[0] + lst[1], lineColor[2], '$z1 + z2$' )
drawline(lst[0]*lst[1], lineColor[3], r'$z1 \times z2$')

plt.axvline(0, color='k', linestyle='--', linewidth=1)
plt.axhline(0, color='k', linestyle='--', linewidth=1)
plt.grid()
plt.axis('square')
plt.xlim(-5, 5)
plt.xlabel('Real Number')
plt.ylim(-5, 5)
plt.ylabel('Imaginary Number')
plt.legend()
plt.show()
```
