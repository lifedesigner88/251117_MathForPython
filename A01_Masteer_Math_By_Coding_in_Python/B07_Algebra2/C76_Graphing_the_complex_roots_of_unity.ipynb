```python
from matplotlib.pyplot import *
from sympy import exp, pi, I, latex, re, im, cos, sin
from IPython.display import display, Math

n = 7

for k in range(n):
    z = exp(2*pi*I*k/n)
    # display(Math(r'\Large (%s)^{%s} \quad \Rightarrow \quad %s'%(latex(z), n, latex(z**n))))
    plot(re(z), im(z), 'o')

axis('square')
xlim(-2, 2)
ylim(-2, 2)
show()
```

```python
from matplotlib.pyplot import *
from sympy import exp, pi, I, re, im
from numpy import linspace

n = 203
line = linspace(0, 0.91, n)

for k in range(n):
    f = line[k]*exp(2*pi*I*k/n)
    plot([0, re(f)], [0, im(f)], c=[line[k], line[k], line[k]], lw=1)

axis('square')
axis('off')
xlim(-1, 1)
ylim(-1, 1)
show()
```
