```python
from matplotlib.pyplot import *
from numpy import arange, zeros
from itertools import product
from time import time

ints = arange(1, 100)

Z = zeros((len(ints), len(ints)))

for x, y in product(ints, ints):
    Z[x-1, y-1] = (x**3 + y**3)**(1/3)

zInts = Z%1 == 0

fig, ax = subplots(2, 2, figsize=(9, 9))
ax = ax.ravel()
ax[0].imshow(Z)
ax[0].set_title(r'$z = \sqrt[3]{x^3 + y^3}$')

ax[1].imshow(zInts)
ax[1].set_title('z = is Int')

for x, y in product(ints, ints):
    Z[x-1, y-1] = (x**2 + y**2)**(1/2)
zInts = Z%1 == 0

ax[2].imshow(Z)
ax[2].set_title(r'$z = \sqrt{x^2 + y^2}$')

ax[3].imshow(zInts)
ax[3].set_title('z = is Int')

for i in ax:
    i.axis('off')

show()
```
