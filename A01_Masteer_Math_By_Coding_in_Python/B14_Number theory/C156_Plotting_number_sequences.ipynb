```python
from matplotlib.pyplot import *
from math import sqrt

nums = range(1, 81)

s1, s2, s3, s4 = zip(*[(
    a/(a+1),
    (a+1)/a,
    a/sum(range(1, a+1)),
    (-1)**a / sqrt(a)
) for a in nums])

dic = {
    "s1": s1,
    "s2": s2,
    "s3": s3,
    "s4": s4
}

figure(figsize=(8, 6))
for key, val in dic.items():
    plot(nums, val, label=key)

legend()
show()
```
