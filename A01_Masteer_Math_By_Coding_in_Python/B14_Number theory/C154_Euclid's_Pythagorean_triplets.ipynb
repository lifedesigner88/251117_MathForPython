```python
from matplotlib.pyplot import *

lst = range(2, 51)

a = []
b = []
c = []

for n in lst:
    for m in lst[n:]:
        a += [m**2 - n**2]
        b += [2*m*n]
        c += [m**2 + n**2]

figure(figsize=(9,6))
scatter(a, b , c=c, marker='o', alpha=.5, lw=.5, ec='k')
axis('off')
show()
```

```python
from itertools import product

lst = range(2, 51)

a, b, c = zip(*[(m**2 - n**2, 2*m*n, m**2 + n**2)
              for m, n in product(lst, lst)])

figure(figsize=(9,6))
scatter(a, b , c=c, marker='o', alpha=.5, lw=.5, ec='k')
axis('off')
show()
```
