```python
from math import sqrt


nums = range(2, 10000 + 1)
answers: list[int] = []

for num in nums:
    divisors: set[int] = {1}
    for divisor in range(2, int(sqrt(num) + 1)):
        if num % divisor == 0:
            divisors.update({divisor, num//divisor})
    if num == sum(divisors):
        answers += [num]
print(answers)
```
