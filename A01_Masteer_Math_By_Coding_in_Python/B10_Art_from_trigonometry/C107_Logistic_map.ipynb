```python
# %matplotlib notebook

from matplotlib.pyplot import *
from numpy import zeros, arange, ones
from matplotlib.colors import hsv_to_rgb

rs = np.linspace(1, 4, 2000)
N = 500
n = 256
h = np.linspace(0, 1, n)
hsv = np.zeros((n, 3))
hsv[:, 0] = h
hsv[:, 1] = 1
hsv[:, 2] = 1
rainbow = hsv_to_rgb(hsv)

x = zeros(N) + 0.5
encap = arange(round(N*0.9), N)
figure(figsize=(9, 16))
for ri in range(len(rs)):
    for n in range(N - 1):
        x[n + 1] = rs[ri]*x[n]*(1 - x[n])
    u = np.unique(x[encap])
    r = rs[ri]*ones(len(u))
    plot(u, -r, '.', markersize=1, color=rainbow[255 - int(((rs[ri] - 1)/3)*255)])
axis('off')
show()
```

```python
n = 256
g = np.linspace(0, 1, n)  # 초록 채널은 0→1

for i in range(len(rainbow)):
    plot(g[i], 1, '.', color=rainbow[i], linewidth=2)
axis('off')
show()
```
