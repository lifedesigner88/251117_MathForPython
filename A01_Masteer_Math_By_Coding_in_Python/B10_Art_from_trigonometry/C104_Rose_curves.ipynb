```python
from matplotlib.pyplot import *
from numpy import cos, sin, pi, linspace

t = linspace(0, 20*pi, 1000)

k = linspace(0, 4, 9)

fig, axs = subplots(3,3, figsize=(10, 10))

axs = axs.ravel()

for i in range(len(k)):
    x = cos(k[i]*t)*cos(t)
    y = cos(k[i]*t)*sin(t)
    axs[i].plot(x, y)
    axs[i].axis("off")

show()
```
