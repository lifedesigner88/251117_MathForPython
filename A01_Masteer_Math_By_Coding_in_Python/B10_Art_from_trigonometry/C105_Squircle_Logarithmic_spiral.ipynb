```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid

x = linspace(-10, 10, 1000)
y = linspace(-10, 10, 1000)

X, Y = meshgrid(x, y)

Fx = X**4 + Y**4

contour(X, Y, Fx, levels=[1, 4, 12, 20])
axis('square')
axis('off')
xlim(-2, 2)
ylim(-2, 2)
```

```python
from matplotlib.pyplot import *
from numpy import linspace, cos, sin, exp, pi

k = 5

t = linspace(0, 13*pi, 3000)

x = cos(t)*exp(t/k)
y = sin(t)*exp(t/k)

x2 = cos(t)*exp(t/k)/2
y2 = sin(t)*exp(t/k)/2

for i in range(len(t)):
    plot([x[i], x2[i]], [y[i], y2[i]], color='k')

axis('equal')
axis('off')
show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, cos, sin, exp, pi


N = 150
t = linspace(0, 4*pi, 1000)
ks = linspace(-6, -2, N)

for i in range(len(ks)):
    x = cos(t)*exp(t/ks[i])
    y = sin(t)*exp(t/ks[i])
    plot(x, y, color=[i/N, i/N, i/N ])

axis('equal')
axis('off')
show()
```
