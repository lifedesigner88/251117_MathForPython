```python
from matplotlib.pyplot import *
from numpy import meshgrid

X, Y = meshgrid(range(10), range(15))

subplot(121)
imshow(X)

subplot(122)
imshow(Y)

show()

subplot(121)
pcolormesh(X, color='k', lw=0.1)

subplot(122)
pcolormesh(Y, color='k', lw=0.1)
```

```python
from numpy import cos, sin, pi, linspace, meshgrid

x = linspace(-pi/2, pi/2, 400)
y = linspace(0, pi, 400)

X, Y = meshgrid(x, y)

Fxy = cos(X) + sin(Y)
imshow(Fxy)
axis('off')
show()
```

```python
from numpy import cos, sin, pi, linspace, meshgrid, exp

x = linspace(-1, 1, 100)
y = linspace(-1, 1, 100)

X, Y = meshgrid(x, y)
Fxy = exp(-(X**2 + Y**2)/1)

imshow(Fxy)
axis('off')
show()
```
