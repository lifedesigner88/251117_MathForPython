```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid

# 중심점 (k, h)
k = 1
h = 1

# 반지름 (a , b)
a = 2
b = 1

x = linspace(-5, 5, 100)
y = linspace(-5, 5, 100)

X, Y = meshgrid(x, y)

Fxy = (X - h)**2/a**2 + (Y - k)**2/b**2

contour(X, Y, Fxy, levels=[1])
axvline(x=0, c='k', ls='--', lw=1)
axhline(y=0, c='k', ls='--', lw=1)

show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid

n = 20
# 중심점 (h, k)
h = 0
k = linspace(8, -8, n)

# 반지름 (a , b)
a = linspace(4, -4, n)
b = 4

x = linspace(-8, 8, 100)
y = linspace(-16, 16, 100)

X, Y = meshgrid(x, y)

for i in range(n):
    Fxy = X**2/a[i]**2 + (Y + k[i])**2/4**2
    contour(X, Y, Fxy, levels=[1], linewidths=1, colors=[i/n, i/n, 0])

axis('square')
xlim(-8, 8)
ylim(-16, 16)
show()
```
