```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid, sqrt

h = 2
k = -3
r = 3

axlim = r + max(abs(h), abs(k))
x = linspace(-axlim, axlim, 100)
y = linspace(-axlim, axlim, 100)

X, Y = meshgrid(x, y)

Fxy = (X - h)**2 + (Y - k)**2 - r**2

subplot(121)
pcolor(X, Y, Fxy)
axis('square')

subplot(122)
contour(X, Y, Fxy, levels=[0])
axis('square')
axvline(0, c='k', ls='--', lw=0.7)
axhline(0, c='k', ls='--', lw=0.7)
plot(h, k, 'ro')
plot(2 + sqrt(5), -1, 'ro')
plot([h, 2 + sqrt(5)], [k, -1], 'r')
show()
```

```python
from sympy import symbols, Eq, solve, Poly

x, y, r = symbols('x, y, r')

rhs = (x - 2)**2 + (y + 3)**2
lhs = 3**2

eq = Eq(rhs, lhs)

display(eq)

fy = solve(eq, x)
display(fy[0].subs({y: -1}))
display(fy[1].subs({y: -1}))
```

```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid

bandwith = 1.5

x = linspace(-5, 5, 100)
y = linspace(-3, 3, 100)
X, Y = meshgrid(x, y)

for i in [-bandwith, +bandwith]:
    Fxy = (X - i)**2 + Y**2
    levels = [radi**2 for radi in linspace(0.5, 3, 15)]
    contour(X, Y, Fxy, levels=levels, colors=[[c/15 for _ in range(3)] for c in range(15)])

plot([-bandwith, +bandwith], [0, 0], 'k', lw=2)
axis("square")
xlim(-6, 6)
ylim(-4, 4)
show()
```
