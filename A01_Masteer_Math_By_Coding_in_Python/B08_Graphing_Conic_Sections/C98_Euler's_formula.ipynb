```python
from matplotlib.pyplot import *
from numpy import pi, exp, cos, sin, angle, abs

k = pi/3
r = 2.3

eul = r*exp(1j*k)
cis = r*cos(k) + r*sin(k)*1j

print(eul)
print(cis)

mag = abs(eul)
ang = angle(eul)
polar([0, ang], [0, mag], 'r')
polar(k, r, 'ko')
show()

from sympy import exp, sin, cos, pi, I, symbols, Eq

theta, r = symbols('theta r')

lhs = r*exp(I*theta)
rhs = r*(cos(theta) + sin(theta)*I)

display(Eq(lhs, rhs))
```

```python
from matplotlib.pyplot import *
from numpy import linspace, sin, cos, pi, abs, arctan2, rad2deg

cospart = 3
sinpart = 4

complex_input = cospart + sinpart*1j
r = abs(complex_input)
ang = arctan2(complex_input.real, complex_input.imag)

print(r)
print(ang)
print(rad2deg(ang))

polar(ang, r, 'ro')
polar([0, ang], [0, r], 'r')
thetagrids([0, 45, 90, 200])
show()
```

```python
from matplotlib.pyplot import *
from numpy import exp, random, pi

n = 300
r = random.rand(n)
theta = random.rand(n)*2*pi

color_list = ("pink", "purple", "green")
alpha_list = (0.8, 0.8, 0.2)

for i in range(n):
    rand_val = i%3
    polar([0, theta[i]], [0, r[i]], color=color_list[rand_val], alpha=alpha_list[rand_val], lw=2)

axis('off')
show()
```

```python
from matplotlib.pyplot import *
from numpy import cos, sin, random, append

n = 400
t = linspace(0, 12*pi, n)

r1, r2 = random.rand(2)

x = cos(r1*t)
y = sin(r2*t)

x = append(x, x[0])
y = append(y, y[0])

plot(x, y)
axis('off')
axis('square')
show()
```
