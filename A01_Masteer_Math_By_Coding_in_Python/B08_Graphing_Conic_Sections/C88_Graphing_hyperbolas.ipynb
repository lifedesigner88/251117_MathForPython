```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid, max

a = 1
b = .5
h = 1
k = 2

axlim = 2*max((a, b)) + max((abs(h), abs(k)))
x = linspace(-axlim, axlim)
X, Y = meshgrid(x, x)

Fxy = (X - h)**2/a**2 - (Y - k)**2/b**2

contour(X, Y, Fxy, levels=[1])
plot(h, k, 'or')
axvline(x=0, c='k', lw=1, ls='--')
axhline(y=0, c='k', lw=1, ls='--')
axis('square')
show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace, meshgrid, max

n = 13

a = linspace(6, 20, n)
b = linspace(6, 20, n)
h = 0
k = 0

axlim = 40
x = linspace(-axlim, axlim, 100)
X, Y = meshgrid(x, x)

for i in range(n):
    Fxy = (X - h)**2/a[i]**2 - (Y - k)**2/b[i]**2
    contour(X, Y, Fxy, levels=[1], colors=[i/n, 0, i/n])
    contour(X, Y, -Fxy, levels=[1], colors=[0, i/n, i/n])

axis('square')
xlim(-50, 50)
ylim(-50, 50)
show()
```
