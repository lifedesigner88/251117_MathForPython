```python
from matplotlib.pyplot import *
from numpy import linspace

## 포물선

a = 1
h = 1
k = -2
n = 100

bandwith = 5

x = linspace(h - bandwith, h + bandwith, n)

y = a*(x - h)**2 - k

plot(x, y)
plot(y, x)
grid()
axis('square')
show()
```

```python
from matplotlib.pyplot import *
from numpy import linspace

h = 2
k = 1
p = .5
a = 1/(4*p)

x = linspace(-4, 6, 100)
y = a*(x - h)**2 + k

plot(x, y, 'y', label='Parabola')
plot(h, k, 'o', label='Vertex')
plot(h, k + p, 'go', label='focus')
axhline(k - p, color='r', label='directrix')
axis('square')
xlim(-6, 6)
ylim(0, 10)
grid()
legend(loc='best')
show()
```

```python
from sympy.abc import x, y, a, h, k
from sympy import Eq, Add, Mul, solve
from sympy.plotting import plot

rhs = a*Mul(Add(x, -h, evaluate=False), Add(x, -h, evaluate=False), evaluate=False)
lhs = y - k
display(Eq(lhs, rhs))

sub = Eq(lhs, rhs).subs({k: 1, h: 2, a: .5})
display(sub)

y_expr = solve(sub, y)[0]
display(y_expr)

p = plot(y_expr, (x, -6, 6), show=False)
p.extend(plot(1, (x, -2, 2), show=False))

p[1].line_color='red'
p.ylim = (0, 10)
p.show()
```
