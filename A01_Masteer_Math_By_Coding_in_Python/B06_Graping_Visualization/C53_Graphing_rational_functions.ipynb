```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3, 4)
y = np.zeros(len(x))

for i in range(len(x)):
    y[i] = 2 - x[i]*x[i]

plt.plot(x, y, 's-')
```

```python
from sympy import simplify, latex, Matrix
from IPython.display import Math


def print_np(np_array):
    display(Math(latex(simplify(np_array))))


x = np.linspace(-3, 3, 1000)
y1 = 2 + np.sqrt(abs(x))
y2 = 2 - np.sqrt(abs(x))

plt.plot(x, y1, 'm')
plt.plot(x, y2, 'k')

plt.show()
```

```python
color = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
powerN = list(range(-1, 4))
x = np.linspace(-4, 4, 1000)

for i in range(len(powerN)):
    y = x ** powerN[i]
    plt.plot(x, y, color[i], label='$ y = x^{%s} $' % powerN[i])

plt.xlim(-4, 4)
plt.ylim(-20, 20)
plt.legend()
plt.show()
```
