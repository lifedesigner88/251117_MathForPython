```python
import matplotlib.pyplot as plt

x = 3
y = 2

plt.plot(x, y, 'o')
plt.axis('square')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.grid()
plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(len(x)):
    plt.plot(x[i], y[i], 'o', label=f"({x[i]}, {y[i]})")

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4,2, 'r+')

axis = plt.gca()  # Get Current Access

ylim = axis.get_ylim() # ylim 은 Y축 리미트 라는 기능.
print(ylim)
axis.set_ylim([ylim[0], 4.333])

plt.show()
```
