```python
import sympy as sp
from sympy.plotting import plot as sp_plot
```

```python
x = sp.symbols('x')

y = x**2

p = sp_plot(y, show=False)
p.xlim = [0, 50]
p[0].line_color = 'm'
p.title = 'this is a nice plot!'
p.show()
```

```python
from sympy import latex

x, a = sp.symbols('x a')

expr = a/x

sub = expr.subs(a, 1)
p = sp_plot(sub, (x, -5, 5), show=False)
p[0].label = '$ y = %s $'%latex(sub)

sub = expr.subs(a, 2)
p1 = sp_plot(sub, (x, -5, 5), show=False)

p.extend(p1)

p[1].label = '$ y = %s $'%latex(sub)

p.ylim = [-5, 5]
p[0].line_color = 'r'
p[1].line_color = 'b'
p.legend = True
p.show()
```

```python
from sympy import symbols, latex
from sympy.plotting import plot

color = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'w']

x, a = symbols('x a')
f = a/(x**2 - a)
p = plot(show=False)

for i in range(4):
    sub = f.subs(a, i + 1)
    p.extend(plot(sub, show=False))
    p[i].line_color = color[i]
    p[i].label = '$ %s $' %latex(sub)

p.xlim = [-5, 5]
p.ylim = [-10, 10]
p.legend = True
p.show()
```
