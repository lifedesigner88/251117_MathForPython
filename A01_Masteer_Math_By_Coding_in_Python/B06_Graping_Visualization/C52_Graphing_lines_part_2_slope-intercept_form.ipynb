```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp



x = [-5, 5]
m = 2  # 기울기
b = 4  # y절편

x = np.array(x)  # x값 범위
y = m * x + b                  # y값 계산

plt.plot(x, y, label='$y = %s $' % y)

plt.axis('square')
plt.grid(True)
plt.axhline(0, color='k',linestyle='--', linewidth=1)  # x축
plt.axvline(0, color='k',linestyle='--', linewidth=1)  # y축
plt.xlim(x)
plt.ylim(x)
plt.legend()
plt.show()
```
