```python
import numpy as np
from matplotlib import pyplot as pp

A = [
    [1, 2],
    [1, 4],
]

pp.imshow(A)

A = np.zeros((10, 14))
print(np.shape(A))

for i in range(np.shape(A)[0]):
    for j in range(np.shape(A)[1]):
        A[i, j] = 3*i - 4*j
        pp.text(j, i, str(int(A[i, j])), ha='center', va='center')

pp.set_cmap('twilight_shifted')
pp.plot([2, 6], [6, 2], 'r')
pp.imshow(A)
pp.savefig('hihihi2.pdf')
```

```python
from matplotlib import pyplot as plt
import numpy as np

n = 10
B = np.zeros((n, n))

for i in range(B.shape[0]):
    for j in range(B.shape[1]):
        B[i, j] = (-1)**(i + j)
plt.set_cmap('binary')
plt.tick_params(labelleft=False, labelbottom=False)
plt.imshow(B)

```
