```python
from matplotlib import pyplot as plt


def init_plt():
    plt.axis('square')
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.grid()
    plt.plot([-6, 6], [0, 0], 'k--', linewidth=.7)
    plt.plot([0, 0], [-6, 6], 'k--', linewidth=.7)


p1 = [-3, 4]
p2 = [-1, 4]

plt.plot(p1, p2, color=[.4, .2, .7], linewidth=3)
init_plt()
plt.show()
```

```python
x = 3
y = 5

plt.plot(x, y, 'ro')
plt.plot([0, x], [0, y], 'r')
init_plt()
```

```python
from sympy import symbols, sqrt, Abs

x = symbols('x')
y = sqrt(Abs(x))

for i in range(-20, 21):
    y_val = y.subs(x, i).evalf(2)
    plt.plot([0, i], [0, y_val])

plt.show()
```

```python

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.axis('square')

a = [0, 0, 2, 2]
b = [0, 2, 2, 0]

#    
plt.plot([a[0], a[1]], [b[0], b[1]], color='red')     #  
plt.plot([a[1], a[2]], [b[1], b[2]], color='green')   #  
plt.plot([a[2], a[3]], [b[2], b[3]], color='blue')    #  
plt.plot([a[3], a[0]], [b[3], b[0]], color='orange')  #  
plt.title("A Colorful square!")

plt.show()
```
