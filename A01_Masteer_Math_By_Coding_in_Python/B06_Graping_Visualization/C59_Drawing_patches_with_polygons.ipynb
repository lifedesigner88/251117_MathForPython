```python
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

y1 = np.array([
    [1, 1], [2, 3], [3, 1]
])
poly = Polygon(y1, fc='m', alpha=0.3, ec='k')

y2 = np.array([
    [2, 2], [2.5, 3.7], [3.5, 1]
])
poly2 = Polygon(y2, facecolor='c', alpha=0.3, edgecolor='k')

plt.gca().add_patch(poly)  # 현재 Axes에 추가
plt.gca().add_patch(poly2)

plt.ylim(0, 4)
plt.xlim(0, 4)
plt.show()

# figure, axis = plt.subplots()
# axis.add_patch(poly)
# axis.add_patch(poly2)
# axis.set_ylim((0, 4))
# axis.set_xlim((0, 4))
```

```python
from matplotlib import pyplot as plot
from matplotlib.patches import Polygon
import numpy as np

x = np.linspace(-2, 2, 100).round(2)
f = (-x**2).round(2)

# dot_list = []
# for i in range(len(x)):
#     dot_list.append([x[i], f[i]])

dot_list = np.vstack((x, f)).T
poly = Polygon(dot_list, facecolor='g', edgecolor='k', alpha=0.3)

square = Polygon([
    [-0.5, -4.0], [-0.5, -2.5], [0.5, -2.5], [0.5, -4.0]
], facecolor='k', edgecolor='k')

plt.gca().add_patch(poly)
plt.gca().add_patch(square)

plot.xlim(-2.2, 2.2)
plot.ylim(-4.2, 0.2)
plot.plot(x, f, 'r')
plot.plot([-2,2], [-4, -4], 'r')
plot.axis('off')

plot.savefig('hihihi2.svg')

plot.show()
```
