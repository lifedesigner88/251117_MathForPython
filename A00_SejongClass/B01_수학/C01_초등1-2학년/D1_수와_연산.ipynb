```python
def read_korean_number(n):
    if not (0 <= n <= 9999):
        return "0~9999 ì‚¬ì´ì˜ ìˆ«ìžë§Œ ì§€ì›ë©ë‹ˆë‹¤."

    units = ["ì¼", "ì‹­", "ë°±", "ì²œ"]
    digits = ["ì˜", "ì¼", "ì´", "ì‚¼", "ì‚¬", "ì˜¤", "ìœ¡", "ì¹ ", "íŒ”", "êµ¬"]

    result = ""
    num_str = str(n).zfill(4)  # 4ìžë¦¬ë¡œ ë§žì¶”ê¸°: 7 â†’ '0007'

    for i in range(4):
        digit = int(num_str[i])
        unit_index = 3 - i
        if digit == 0:
            continue
        if digit == 1:
            result += units[unit_index]
        else: # 2 ~ 9
            if unit_index == 0:
                result += digits[digit]
            else:
                result += digits[digit] + units[unit_index]

    return result if result else digits[0]


def parse_korean_number(text):
    digits = ["ì˜", "ì¼", "ì´", "ì‚¼", "ì‚¬", "ì˜¤", "ìœ¡", "ì¹ ", "íŒ”", "êµ¬"]
    result = ["0", "0", "0", "0"]  # ì²œ, ë°±, ì‹­, ì¼ ìžë¦¬
    units = ["ì²œ", "ë°±", "ì‹­"]

    for unit_index, unit in enumerate(units):
        if unit in text:
            idx = text.index(unit)
            if idx > 0 and text[idx - 1] in digits:
                result[unit_index] = str(digits.index(text[idx - 1]))
            else:
                result[unit_index] = "1"

    # ì¼ì˜ ìžë¦¬ ì²˜ë¦¬: ë§ˆì§€ë§‰ ê¸€ìžê°€ ìˆ«ìžê°’ ì´ë©´ ì¼ì˜ ìžë¦¬
    if text and text[-1] in digits:
        result[3] = str(digits.index(text[-1]))

    return int("".join(result))



from random import randint
nums = [randint(0, 100) for _ in range(5)]
nums = nums + [randint(100, 1000) for _ in range(7)]
nums = nums + [randint(1000, 10000) for _ in range(8)] + [0]

# ì˜ˆì‹œ ì¶œë ¥
for num in nums:
    text = read_korean_number(num)
    result = parse_korean_number(text)
    is_same = num == result
    print(f"{num:04d} â†’ {text} -> {result} , {"âœ…" if is_same else "ðŸš¨"}")


from itertools import product
combinations = list(product([0, 1], repeat=4))
numbers = [int("".join(map(str, c))) for c in combinations]

# ì¶œë ¥
for num in numbers:
    text = read_korean_number(num)
    result = parse_korean_number(text)
    is_same = num == result
    print(f"{num:04d} â†’ {text} -> {result} , {"âœ…" if is_same else "ðŸš¨"}")
```
