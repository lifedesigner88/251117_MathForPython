```python
import numpy as np
from matplotlib.pyplot import *
from mpl_toolkits.mplot3d import Axes3D  # noqa


AMPLITUDE = 0.5  #    (0.3 ~ 0.7  )

#   ( )
def f(x, y):
    return AMPLITUDE * np.sin(x) * np.cos(y)

#  
x = np.linspace(-3, 3, 120)
y = np.linspace(-3, 3, 120)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection="3d")

# 
ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.85, linewidth=0)

# ---    2 () ---
p1 = (np.random.uniform(-2, 2), np.random.uniform(-2, 2))
p2 = (np.random.uniform(-2, 2), np.random.uniform(-2, 2))

ax.scatter(*p1, f(*p1), color="red", s=50)
ax.scatter(*p2, f(*p2), color="red", s=50)

# ---    2 ( ) ---
x_line = np.linspace(-3, 3, 200)
y_line = np.linspace(-3, 3, 200)

y_const = np.random.uniform(-2, 2)
ax.plot(x_line, y_const, f(x_line, y_const),
        color="red", linewidth=2)

x_const = np.random.uniform(-2, 2)
ax.plot(x_const, y_line, f(x_const, y_line),
        color="red", linewidth=2)

# ---  /   ---
ax.set(xticks=[], yticks=[], zticks=[])
ax.grid(False)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_zlabel("")
ax.view_init(elev=40, azim=50)

for axis in (ax.xaxis, ax.yaxis, ax.zaxis):
    axis.pane.set_visible(False)
tight_layout()
show()

```
